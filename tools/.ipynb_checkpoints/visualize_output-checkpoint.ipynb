# ⚠️ 重要更新 (2026-01-20): P0-1 门控修正、Selection Bias 风险与双口径报告

## 背景与问题识别

历史上 `layout` 评估曾引入 **`n_pairs_mismatch` 直接 gate**，这会带来明显的 **selection bias 风险**：
- 一批“强纠错/高工作量”的样本（角点数变化大）被系统性排除
- 主结论可能只来自 `layout_used=True` 的子集（覆盖率不稳定）
- 审稿人会质疑代表性与缺失机制（MNAR）

## P0-1（已采纳的解决策略）

把“结构差异”从 **硬门控** 降级为 **分层/诊断信号**：
- **标准 layout 指标（polygon/深度类）**：不再要求 pred/ann 角点对数一致；不因 mismatch 失败
- **点级可比指标（pointwise RMSE）**：仍保留可比性门控（需要稳定匹配/覆盖率）
- **Δn_pairs / odd points / pairing warnings**：作为分层变量，用于解释工作量、筛选反例与任务分配（RQ3）

## 双口径报告（论文必须固定输出）

- **A 口径：标准 Layout 指标**
  - 数据集：`df_main`（默认 in-scope 且 layout 指标可计算）
  - 指标：layout_2d_iou, layout_3d_iou, layout_depth_rmse 等
  - 用途：与 HoHoNet/HorizonNet 对标
- **B 口径：结构无关指标（全 In-scope）**
  - 数据集：`df_inscope`（in-scope，不强制 layout_used）
  - 指标：boundary_rmse_px, active_time, edit_magnitude, |Δn_pairs| 等
  - 用途：覆盖全部有效标注（含强纠错样本），避免“只报能算的样本”

## 本 notebook 的组织方式

- 先跑一次清洗，生成 `df_transparent/df_main/df_rel`（统一口径）
- 论文主图/表放在前面；其余诊断图放到“附录/审计”部分

````python
# =============================================================================
# Section 0.3: 公共工具函数（消除冗余）
# =============================================================================

# -----------------------------------------------------------------------------
# A. 数据类型转换工具
# -----------------------------------------------------------------------------
def to_numeric_safe(s: pd.Series, errors: str = "coerce") -> pd.Series:
    """安全转换为数值类型，统一处理 NA/异常值。"""
    return pd.to_numeric(s, errors=errors)

def to_bool_tristate(x) -> Optional[bool]:
    """稳健的三态布尔：True / False / None（未知/缺失保持 None）。"""
    if pd.isna(x):
        return None
    if isinstance(x, (bool, np.bool_)):
        return bool(x)
    s = str(x).strip().lower()
    if s in {"true", "1", "yes", "y", "t"}:
        return True
    if s in {"false", "0", "no", "n", "f"}:
        return False
    return None

def clean_str_col(df: pd.DataFrame, col: str) -> None:
    """标准化字符串列：去空格、小写、空串/伪空值 → NA。"""
    if col not in df.columns:
        return
    s = df[col].astype("string")
    s = s.str.strip().str.lower()
    s = s.replace({"": pd.NA, "nan": pd.NA, "none": pd.NA})
    df[col] = s

# -----------------------------------------------------------------------------
# B. 统计计算工具
# -----------------------------------------------------------------------------
def safe_median(s: pd.Series) -> Optional[float]:
    """安全计算中位数，过滤 NA。"""
    vals = to_numeric_safe(s).dropna()
    return float(vals.median()) if len(vals) > 0 else None

def safe_mean(s: pd.Series) -> Optional[float]:
    """安全计算均值，过滤 NA。"""
    vals = to_numeric_safe(s).dropna()
    return float(vals.mean()) if len(vals) > 0 else None

def safe_percentile(s: pd.Series, q: float) -> Optional[float]:
    """安全计算分位数。"""
    vals = to_numeric_safe(s).dropna()
    return float(np.percentile(vals, q)) if len(vals) > 0 else None

def bootstrap_ci(
    values: pd.Series,
    stat_fn: Callable = np.median,
    n_iters: int = None,
    ci: float = None,
    seed: int = None,
) -> tuple:
    """Percentile bootstrap CI。返回 (stat, ci_low, ci_high)。"""
    n_iters = n_iters or CONFIG.get("bootstrap_n", 1000)
    ci = ci or CONFIG.get("bootstrap_ci", 0.95)
    seed = seed or CONFIG.get("random_seed", 42)
    
    vals = to_numeric_safe(values).dropna().to_numpy()
    if len(vals) == 0:
        return (None, None, None)
    
    stat = float(stat_fn(vals))
    if len(vals) == 1:
        return (stat, stat, stat)
    
    alpha = 0.5 * (1.0 - ci)
    rng = np.random.default_rng(seed)
    boot = np.array([stat_fn(rng.choice(vals, size=len(vals), replace=True)) for _ in range(n_iters)])
    ci_low, ci_high = np.percentile(boot, [100 * alpha, 100 * (1 - alpha)])
    return (stat, float(ci_low), float(ci_high))

def print_dataset_stats(df: pd.DataFrame, name: str) -> None:
    """统一打印数据集规模，便于审稿核对口径一致性。"""
    parts = [f"name={name}", f"n_rows={len(df)}"]
    if "task_id" in df.columns:
        parts.append(f"n_tasks={df['task_id'].nunique()}")
    if "title" in df.columns:
        parts.append(f"n_titles={df['title'].nunique()}")
    if "annotator_id" in df.columns:
        parts.append(f"n_annotators={df['annotator_id'].nunique()}")
    print("📌 数据集规模: " + ", ".join(parts))

# -----------------------------------------------------------------------------
# C. Condition/Scope 转换工具（统一入口，消除重复）
# -----------------------------------------------------------------------------
CONDITION_LABELS = {
    "manual": "Manual",
    "semi": "Semi-Auto",
    "unknown": "Unknown",
}

SCOPE_OOS_PREFIXES = ("oos", "out-of-scope", "out of scope")
SCOPE_NORMAL_VALUES = ("normal", "in-scope", "camera room")

def normalize_condition(s: str) -> str:
    """统一 condition 字段格式。"""
    if pd.isna(s):
        return "unknown"
    return str(s).strip().lower()

def condition_display(s: str) -> str:
    """获取 condition 的显示标签。"""
    return CONDITION_LABELS.get(normalize_condition(s), str(s))

def is_oos_scope(s) -> bool:
    """判断 scope 是否为 OOS。"""
    if pd.isna(s):
        return False
    ss = str(s).strip().lower()
    return any(ss.startswith(prefix) for prefix in SCOPE_OOS_PREFIXES)

def is_normal_scope(s) -> bool:
    """判断 scope 是否为 normal/in-scope。"""
    if pd.isna(s):
        return False
    ss = str(s).strip().lower()
    return ss == "normal" or any(v in ss for v in SCOPE_NORMAL_VALUES)

# -----------------------------------------------------------------------------
# D. 图表保存工具
# -----------------------------------------------------------------------------
def save_figure(
    fig: plt.Figure,
    name: str,
    category: str = "diagnostics",
    formats: list = None,
    dpi: int = 150,
) -> list:
    """
    统一的图片保存接口。
    
    Parameters:
        fig: matplotlib Figure 对象
        name: 图片名称（不含扩展名）
        category: 保存的子目录类别（对应 CONFIG["fig_dirs"] 的 key）
        formats: 保存格式列表，默认使用 CONFIG["figure_format"]
        dpi: 分辨率
    
    Returns:
        保存的文件路径列表
    """
    if not CONFIG.get("save_figures", True):
        return []
    
    formats = formats or [CONFIG.get("figure_format", "pdf")]
    if isinstance(formats, str):
        formats = [formats]
    
    save_dir = CONFIG["fig_dirs"].get(category, CONFIG["output_dir"])
    saved_paths = []
    
    for fmt in formats:
        filepath = save_dir / f"{name}.{fmt}"
        fig.savefig(filepath, dpi=dpi, bbox_inches="tight", facecolor="white")
        saved_paths.append(filepath)
        print(f"💾 [saved] {filepath}")
    
    return saved_paths

def save_table(
    df: pd.DataFrame,
    name: str,
    category: str = "tables",
    index: bool = True,
) -> Path:
    """统一的表格保存接口。"""
    save_dir = CONFIG["fig_dirs"].get(category, CONFIG["output_dir"])
    filepath = save_dir / f"{name}.csv"
    df.to_csv(filepath, index=index)
    print(f"💾 [saved] {filepath}")
    return filepath

print("✅ 公共工具函数定义完成")
````

````python
# =============================================================================
# Section 0.4: 统一数据加载与条件校验 (Unified Flow)
# =============================================================================

def load_and_verify_all(manifest: dict, analysis_dir: Path) -> pd.DataFrame:
    """
    根据清单加载多组 CSV，自动注入 dataset_group 并校验数据完整性与属性一致性。
    """
    all_dfs = []
    
    for group_name, info in manifest.items():
        csv_path = analysis_dir / info["file"]
        if not csv_path.exists():
            print(f"⚠️ [Skip] 找不到文件: {info['file']} (Group: {group_name})")
            continue
            
        df = pd.read_csv(csv_path)
        
        # 1. 自动填补缺少的 dataset_group (向下兼容)
        if "dataset_group" not in df.columns:
            df["dataset_group"] = group_name
            
        # 2. 提取当前分组数据
        group_df = df[df["dataset_group"] == group_name].copy()
        
        if len(group_df) == 0:
            print(f"⚠️ [Warning] 分组 {group_name} 在 CSV 中无记录")
            continue

        # 3. 属性一致性校验 (Manifest vs Physical Data)
        if "condition" in group_df.columns:
            # 统计实际数据中的 condition
            counts = group_df["condition"].value_counts()
            dominant_condition = counts.index[0]
            
            # 验证与 Manifest 中的预期是否一致
            expected_cond = info.get("condition")
            if expected_cond and dominant_condition != expected_cond:
                print(f"❌ [LABEL MISMATCH] 分组 '{group_name}' 在 Manifest 中定义为 {expected_cond}，"
                      f"但数据物理检测为 {dominant_condition}！")
                print(f"   原因: 用于该组的 JSON 导出文件可能包含预测信息(semi)或不含预测(manual)。")
        
        # 4. 注入元数据
        group_df["analysis_role"] = info.get("role", "unknown")
        all_dfs.append(group_df)
        print(f"✅ Loaded {group_name}: {len(group_df)} rows | Role: {info.get('role')} | Data: {dominant_condition if 'condition' in group_df.columns else 'N/A'}")

    if not all_dfs:
        raise ValueError("严重错误: MANIFEST 中定义的所有分组均未在 CSV 中找到有效数据。")
        
    full_df = pd.concat(all_dfs, ignore_index=True)
    return full_df

# 执行加载
df_raw = load_and_verify_all(MANIFEST, CONFIG["analysis_dir"])

print(f"\n📊 纵向汇总完成: 共 {df_raw.shape[0]} 条记录")
print(f"实验组分布:\n{df_raw['dataset_group'].value_counts()}")
````

# Section 1: Data Cleaning & Definitions

- 统一字段语义、掩码口径
- 生成 `df_transparent / df_main / df_rel`

### 1.1 核心数据结构与筛选逻辑 (2026-01-26 Updated)

为确保实验分析的严谨性，并在复现时消除**选择性偏差 (Selection Bias)**，我们对数据筛选逻辑进行了以下重要修正：

*   **df_main (Performance Analysis)**:
    *   **Scope**: 仅包含 `In-Scope` (正常视野) 的任务。
    *   **Metrics**: 适用于 `IoU`、`Active Time`、`Boundary RMSE` 等 2D 指标。
    *   **Logic Change**: **不再强制 `layout_used=True`**。
        *   *原因*: `layout_used` 依赖于角点配对算法 (pair_keypoints)，若模型预测角点数与 GT 不一致会导致配对失败。如果因此剔除这些样本，会人为抬高 IoU 平均值（因为被剔除的往往是表现较差的样本），造成 "Cherry-picking" 的假象。
        *   *新策略*: 此子集包含所有 `In-Scope` 样本，即使配对失败，其 Polygons IoU 依然有效且应被计入。

*   **df_layout_3d (Layout Specific)**:
    *   **Scope**: `df_main` 的子集。
    *   **Metrics**: 适用于 `3D IoU`、`Depth RMSE`、`Delta_1`。
    *   **Logic**: 必须满足 `layout_used=True`。只有成功重建曼哈顿布局的样本才能计算 3D 指标。

*   **df_rel (Reliability Analysis)**:
    *   **Scope**: 仅包含多标任务 (Annotators ≥ 3)。
    *   **Metrics**: `r_u` (Leave-one-out reliability)。
    *   **Logic**: 必须包含专家共识分 (`iou_to_consensus_loo`)。

---

### 1.2 统一数据流 (Unified Flow) 说明

为了简化 A/B 测试 (Manual vs Semi) 的流程，我们引入了“统一CSV”模式：
1.  **数据源**: 所有实验组（Manual, Semi, Calibration）通过由 `analyze_quality.py --append` 生成的单个 CSV (`quality_report_YYYYMMDD.csv`) 加载。
2.  **唯一标识**: 使用 `dataset_group` (实验组) 和 `title` (图像文件名) 的组合作为唯一键。
    *   *优势*: 即使 `task_id` 在不同项目中不一致，只要图像文件名相同，即可自动进行 Paired T-Test。
3.  **安全性**: `clean_quality_df` 会自动处理这一合并，并根据 MANIFEST 提取所需子集。


````python
# =============================================================================
# Section 1: 数据清洗与口径定义 (Unified Flow Edition)
# =============================================================================

def clean_quality_df(
    df: pd.DataFrame,
    require_layout_used_for_main: bool = True,
    verbose: bool = True,
) -> pd.DataFrame:
    r"""
    清洗 analyze_quality.py 导出的长表。
    2026-01-25 升级: 增加分析角色隔离 (Performance vs Reliability)
    2026-01-26 升级: 强化 'title' 字段标准化，以支持跨组 (Manual vs Semi) 的配对分析
                     修复 n_pairs_mismatch 引入的筛选偏差 (Selection Bias)
    """
    df = df.copy()
    
    # 1) 基础标准化 (增加识别 title，确保跨项目配对的唯一键)
    id_cols = ["task_id", "annotator_id", "dataset_group", "analysis_role", "title", "image_name"]
    for col in id_cols:
        if col in df.columns:
            df[col] = df[col].astype(str)
    
    # Alias: 确保 'title' 是图像的通用标识符
    if "image_name" in df.columns and ("title" not in df.columns or df["title"].iloc[0] == "nan"):
        df["title"] = df["image_name"]
            
    for col in ["scope", "condition", "layout_gate_reason"]:
        clean_str_col(df, col)

    # 2) 语义识别（保持 scope_clean 与 scope_missing）
    df["scope_clean"] = df["scope"] if "scope" in df.columns else pd.NA
    df["scope_missing"] = df["scope_clean"].isna()
    df["is_oos_clean"] = df["scope_clean"].apply(is_oos_scope)
    df["is_normal_clean"] = df["scope_clean"].apply(is_normal_scope)
    df["condition_clean"] = df["condition"].apply(normalize_condition)
    
    # 3) BOOL 转换
    for col in ["layout_used", "reliability_used", "pointwise_rmse_used"]:
        if col in df.columns:
            df[f"{col}_clean"] = df[col].apply(to_bool_tristate)

    # 4) 指标安全性转换
    num_cols = ["iou", "iou_3d", "active_time", "boundary_rmse_px", "iou_to_consensus_loo"]
    for col in num_cols:
        if col in df.columns:
            df[col] = to_numeric_safe(df[col])

    # 5) 定义有效性掩码 (核心逻辑修复点)
    df["data_valid_inscope"] = df["is_normal_clean"] & (~df["scope_missing"])
    
    # [FIXED] 消除选择性偏差 (Selection Bias):
    # 原逻辑: 要求 layout_used 为真，这会剔除所有由于角点失配 (n_pairs_mismatch) 的合法标注。
    # 新逻辑: 对于 IoU 和 Active Time 分析，只需 In-scope 即可。 
    #         只有 3D 深度/RMSE 等依赖点对齐的指标才受 layout_used 约束。
    
    df["data_valid_for_main"] = (
        (df["analysis_role"] == "performance")
        & df["data_valid_inscope"]
        # & (df["layout_used_clean"] == True)  # <-- 此处移除强耦合，解决偏差
    )
    
    # 专门保留一个用于 3D 指标的子集
    df["data_valid_for_layout_3d"] = (
        df["data_valid_for_main"] & (df["layout_used_clean"] == True)
    )
    
    # 可靠性掩码 (Reliability): 必须是 reliability 角色 + 有专家分 + 多人标注
    if "iou_to_consensus_loo" in df.columns:
        df["data_valid_for_reliability"] = (
            (df["analysis_role"] == "reliability")
            & df["data_valid_inscope"]
            & df["iou_to_consensus_loo"].notna()
            & (df["reliability_used_clean"] == True)
        )
    else:
        df["data_valid_for_reliability"] = False

    if verbose:
        print(f"Total rows: {len(df)}")
        print(f"Main (Perf) subset: {df['data_valid_for_main'].sum()}")
        print(f"Layout 3D subset: {df['data_valid_for_layout_3d'].sum()}")
        print(f"Reliability subset: {df['data_valid_for_reliability'].sum()}")
        print(f"Unique images (Titles): {df['title'].nunique()}")
        
        if df["data_valid_for_reliability"].sum() == 0:
            print("⚠️ Reliability subset 为 0：当前数据可能缺少 reliability 角色或 Calibration_* 数据。")
            if "analysis_role" in df.columns:
                print("   analysis_role 分布:", df["analysis_role"].value_counts(dropna=False).to_dict())
            if "reliability_used_clean" in df.columns:
                print("   reliability_used_clean=True:", int((df["reliability_used_clean"] == True).sum()))
            if "iou_to_consensus_loo" in df.columns:
                print("   iou_to_consensus_loo 非空:", int(df["iou_to_consensus_loo"].notna().sum()))
            print("   建议：对 Calibration_* 数据运行 analyze_quality.py --analysis_role reliability --append")
        
    return df

# 执行清洗并生成视图
df_clean = clean_quality_df(df_raw, verbose=True)

# 核心视图隔离
df_transparent = df_clean
df_main = df_clean[df_clean["data_valid_for_main"]].copy()
df_rel  = df_clean[df_clean["data_valid_for_reliability"]].copy()

# 可靠性子集按校准组拆分（用于严格区分 r_u 与 RQ2）
if "dataset_group" in df_rel.columns:
    df_rel_manual = df_rel[df_rel["dataset_group"] == "Calibration_manual"].copy()
    df_rel_semi = df_rel[df_rel["dataset_group"] == "Calibration_semi"].copy()
else:
    df_rel_manual = df_rel.copy()
    df_rel_semi = df_rel.iloc[0:0].copy()

print(f"\n✅ 视图生成完成: df_main({len(df_main)}), df_rel({len(df_rel)})")
print_dataset_stats(df_rel_manual, "df_rel_manual (Calibration_manual)")
print_dataset_stats(df_rel_semi, "df_rel_semi (Calibration_semi)")
````

# Section 2: Transparency & Data Hygiene

- 覆盖率与门控原因（P0-1）
- 表 P0 / 图 P1-P3 相关输出

## [论文图 Fig.P3 | 透明披露 / P0-1] 覆盖率与门控原因（Selection Bias 检查）

**放置位置**：主文 Section 4（透明披露），并在 P0-1 讨论处引用。
**回答的问题**：你的主结果是否因门控造成系统性缺失（MNAR/selection bias）？mismatch 子群是否代表“高工作量/强纠错”而被系统性排除？

**数据集**：`df_transparent`（全量）与 `df_inscope`（in-scope 子集）

**最少需要输出**（建议做成一个 figure 的多 panel）：
- Panel A：in-scope 覆盖率 `mean(layout_used_clean)`（并列展示 manual vs semi）
- Panel B：`layout_gate_reason` 在 in-scope 且 layout_used=False 条件下的分布（Top-K + 其余合并）
- Panel C：mismatch 子群 vs 可计算子群 的 active_time / BoundaryRMSE / |Δn_pairs| 对比（核心证据：高工作量子群被剔除）
- Panel D（可选）：按 annotator 的覆盖率差异（检查缺失机制是否与人耦合）

````python
# =============================================================================
# [论文图 Fig.P3] 覆盖率与门控透明性报告（多 panel 可视化）
# =============================================================================
# 数据集：df_transparent（in-scope 子集）
# 目的：透明披露 selection bias 风险与 mismatch 子群工作量
# =============================================================================

def plot_coverage_and_gate_transparency(df: pd.DataFrame) -> plt.Figure:
    """绘制覆盖率与门控透明性分析图（Fig.P3）。"""
    
    df_inscope = df[df['data_valid_inscope']].copy()
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # ========== Panel A: 覆盖率统计 ==========
    ax1 = axes[0, 0]
    coverage_rate = df_inscope['layout_used_clean'].mean()
    n_used = int(df_inscope['layout_used_clean'].sum())
    n_total = len(df_inscope)
    
    colors = ['#55A868', '#C44E52']
    counts = [n_used, n_total - n_used]
    labels = [f'Layout Used\n({n_used}, {coverage_rate:.1%})', 
              f'Gated\n({n_total - n_used}, {1-coverage_rate:.1%})']
    
    wedges, texts, autotexts = ax1.pie(counts, labels=labels, colors=colors, autopct='',
                                       startangle=90, textprops={'fontsize': 11, 'weight': 'bold'})
    ax1.set_title(f"Panel A: Coverage Rate (In-scope n={n_total})", fontweight='bold')
    
    # 添加警告标注
    if coverage_rate < 0.7:
        ax1.text(0, -1.3, "Selection Bias Risk: Coverage < 70%", 
                ha='center', fontsize=10, color='red', weight='bold',
                bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))
    
    # ========== Panel B: 门控原因分布 ==========
    ax2 = axes[0, 1]
    df_gated = df_inscope[~df_inscope['layout_used_clean']]
    
    if len(df_gated) > 0:
        gate_dist = df_gated['layout_gate_reason'].value_counts()
        
        # 颜色：n_pairs_mismatch 用红色标记（P0-1 回归检测）
        bar_colors = ['#C44E52' if 'mismatch' in str(r).lower() else '#4C72B0' 
                     for r in gate_dist.index]
        
        ax2.barh(range(len(gate_dist)), gate_dist.values, color=bar_colors)
        ax2.set_yticks(range(len(gate_dist)))
        ax2.set_yticklabels(gate_dist.index)
        ax2.set_xlabel('Count')
        ax2.set_title(f"Panel B: Gate Reasons (n={len(df_gated)})", fontweight='bold')
        ax2.invert_yaxis()
        
        # 添加百分比标签
        for i, (reason, count) in enumerate(gate_dist.items()):
            pct = count / len(df_gated)
            ax2.text(count + 0.1, i, f"{pct:.1%}", va='center', fontsize=9)
        
        # P0-1 校验
        if 'n_pairs_mismatch' in gate_dist.index:
            ax2.text(0.5, -0.15, "REGRESSION: n_pairs_mismatch still exists!",
                    transform=ax2.transAxes, ha='center', color='red', weight='bold',
                    fontsize=9, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))
    else:
        ax2.text(0.5, 0.5, 'No gated samples', ha='center', va='center', transform=ax2.transAxes)
        ax2.axis('off')
    
    # ========== Panel C: Mismatch vs Matched 工作量对比 ==========
    ax3 = axes[1, 0]
    
    if all(c in df_inscope.columns for c in ['pred_n_pairs', 'ann_n_pairs', 'active_time']):
        pred_pairs = pd.to_numeric(df_inscope['pred_n_pairs'], errors='coerce')
        ann_pairs = pd.to_numeric(df_inscope['ann_n_pairs'], errors='coerce')
        delta_pairs = pred_pairs - ann_pairs
        
        df_inscope['_is_mismatch'] = (delta_pairs.notna() & delta_pairs.ne(0)).map(
            {True: 'Mismatch\n(Delta-n!=0)', False: 'Matched\n(Delta-n=0)'})
        df_inscope['_active_time'] = pd.to_numeric(df_inscope['active_time'], errors='coerce')
        
        df_plot = df_inscope.dropna(subset=['_is_mismatch', '_active_time'])
        
        if len(df_plot) > 0:
            order = ['Matched\n(Delta-n=0)', 'Mismatch\n(Delta-n!=0)']
            sns.boxplot(data=df_plot, x='_is_mismatch', y='_active_time', order=order,
                       palette={'Matched\n(Delta-n=0)': '#55A868', 'Mismatch\n(Delta-n!=0)': '#C44E52'},
                       ax=ax3)
            ax3.set_xlabel('')
            ax3.set_ylabel('Active Time (seconds)')
            ax3.set_title('Panel C: Workload by Structural Difference', fontweight='bold')
            
            # 添加统计标注
            for i, status in enumerate(order):
                vals = df_plot.loc[df_plot['_is_mismatch'] == status, '_active_time']
                if len(vals) > 0:
                    med = vals.median()
                    ax3.text(i, med, f'n={len(vals)}\nmed={med:.0f}s',
                            ha='center', va='bottom', fontsize=9, weight='bold')
    else:
        ax3.text(0.5, 0.5, 'Insufficient data', ha='center', va='center', transform=ax3.transAxes)
        ax3.axis('off')
    
    # ========== Panel D: 按 Condition 覆盖率（如果有多条件） ==========
    ax4 = axes[1, 1]
    
    if 'condition_clean' in df_inscope.columns:
        cond_stats = df_inscope.groupby('condition_clean').agg({
            'layout_used_clean': ['sum', 'count', 'mean']
        }).reset_index()
        cond_stats.columns = ['Condition', 'n_used', 'n_total', 'coverage']
        cond_stats['Condition'] = cond_stats['Condition'].apply(condition_display)
        
        x_pos = range(len(cond_stats))
        bars = ax4.bar(x_pos, cond_stats['coverage'] * 100, 
                      color=['#4C72B0', '#55A868', '#999999'][:len(cond_stats)])
        ax4.set_xticks(x_pos)
        ax4.set_xticklabels(cond_stats['Condition'])
        ax4.set_ylabel('Coverage Rate (%)')
        ax4.set_title('Panel D: Coverage by Condition', fontweight='bold')
        ax4.set_ylim(0, 105)
        
        # 添加数值标签
        for i, row in cond_stats.iterrows():
            ax4.text(i, row['coverage'] * 100 + 2, 
                    f"{row['coverage']:.1%}\n({int(row['n_used'])}/{int(row['n_total'])})",
                    ha='center', va='bottom', fontsize=9)
    else:
        ax4.text(0.5, 0.5, 'No condition variable', ha='center', va='center', transform=ax4.transAxes)
        ax4.axis('off')
    
    plt.suptitle('Fig.P3: Coverage, Gate Transparency & Selection Bias Check', 
                fontsize=14, fontweight='bold', y=0.995)
    plt.tight_layout()
    
    return fig

# 绘制并保存
fig_p3 = plot_coverage_and_gate_transparency(df_transparent)
save_figure(fig_p3, "Fig_P3_coverage_gate_transparency", category="transparency")
plt.show()

# 打印关键统计（补充文本摘要）
df_inscope = df_transparent[df_transparent['data_valid_inscope']].copy()
print("\n" + "=" * 70)
print("📊 Fig.P3 关键统计摘要")
print("=" * 70)
print(f"In-scope 样本数: {len(df_inscope)}")
print(f"Layout Used: {df_inscope['layout_used_clean'].sum()} ({df_inscope['layout_used_clean'].mean():.1%})")

# P0-1 校验
df_gated = df_inscope[~df_inscope['layout_used_clean']]
if len(df_gated) > 0:
    gate_dist = df_gated['layout_gate_reason'].value_counts()
    if 'n_pairs_mismatch' in gate_dist.index:
        print(f"\n🔴 REGRESSION: {gate_dist['n_pairs_mismatch']} samples gated by n_pairs_mismatch")
    else:
        print("\n✅ P0-1 校验通过：无 n_pairs_mismatch gate")

# Mismatch 子群证据
if all(c in df_inscope.columns for c in ['pred_n_pairs', 'ann_n_pairs', 'active_time']):
    pred_pairs = pd.to_numeric(df_inscope['pred_n_pairs'], errors='coerce')
    ann_pairs = pd.to_numeric(df_inscope['ann_n_pairs'], errors='coerce')
    delta = pred_pairs - ann_pairs
    
    mismatch_mask = delta.notna() & delta.ne(0)
    matched_mask = delta.notna() & delta.eq(0)
    
    if mismatch_mask.any() and matched_mask.any():
        m_time = df_inscope.loc[mismatch_mask, 'active_time'].median()
        t_time = df_inscope.loc[matched_mask, 'active_time'].median()
        print(f"\nMismatch 子群: n={mismatch_mask.sum()}, active_time med={m_time:.0f}s")
        print(f"Matched 子群:  n={matched_mask.sum()}, active_time med={t_time:.0f}s")
        if m_time > t_time:
            print("  → Mismatch 子群耗时更高（强纠错证据）")
print("=" * 70)
````

````python
# [可选补充 | RQ2] layout-rendered depth 指标摘要（不参与 3D IoU 计算）
# 目的：回答“省时是否伴随几何偏差变差”的旁证；并披露 depth 指标缺失率（depth_failed）

df_depth = df_main.copy()  # 主口径：in-scope +（可选）layout_used=True
for col in ['layout_depth_rmse', 'layout_delta1', 'layout_3d_iou']:
    if col in df_depth.columns:
        df_depth[col] = pd.to_numeric(df_depth[col], errors='coerce')

if 'condition_clean' not in df_depth.columns:
    print("⚠️  缺少 condition_clean，跳过 depth 指标分条件汇总")
else:
    rows = []
    for cond in ['manual', 'semi']:
        sub = df_depth[df_depth['condition_clean'] == cond]
        if len(sub) == 0:
            continue
        rmse = sub.get('layout_depth_rmse')
        d1 = sub.get('layout_delta1')
        rows.append({
            'condition': cond,
            'n': int(len(sub)),
            'depth_rmse_available_rate': float(rmse.notna().mean()) if rmse is not None else np.nan,
            'depth_rmse_med': float(rmse.median()) if rmse is not None and rmse.notna().any() else np.nan,
            'delta1_available_rate': float(d1.notna().mean()) if d1 is not None else np.nan,
            'delta1_med': float(d1.median()) if d1 is not None and d1.notna().any() else np.nan,
        })

    depth_summary = pd.DataFrame(rows)
    display(depth_summary)

    if 'CONFIG' in globals():
        out_path = CONFIG['output_dir'] / 'table_depth_metrics_by_condition.csv'
        depth_summary.to_csv(out_path, index=False)
        print(f"[saved] {out_path}")

# 额外诊断：depth_rmse 与 3d_iou 的相关性（仅作 sanity check）
if all(c in df_depth.columns for c in ['layout_depth_rmse', 'layout_3d_iou']):
    tmp = df_depth[['layout_depth_rmse', 'layout_3d_iou']].dropna()
    if len(tmp) >= 10:
        corr = float(tmp['layout_depth_rmse'].corr(tmp['layout_3d_iou'], method='spearman'))
        print(f"Spearman corr(depth_rmse, 3d_iou): {corr:.3f} (n={len(tmp)})")
````

## [论文表 Table.P0 | 口径审计/透明披露] 字段缺失率与 active_time 覆盖率

**放置位置**：主文 Section 4（透明披露/数据卫生）
**目的**：让审稿人明确：哪些字段缺失、缺失是否系统性、耗时记录是否可靠（active_time 的 0/缺失/长尾），以及 gate 字段是否存在“静默丢样本”。

**数据集**：`df_transparent`（无过滤）

**核心列**：
- `scope_clean`/`scope_missing`
- `difficulty`, `model_issue`（以及 *_reported 统计）
- `active_time`（含 active_time==0 比例）
- `active_time_source`（active_logs/lead_time/missing）
- `active_time_display`（用于表格展示：lead_time 会带括号说明）
- gate：`layout_used_clean`, `pointwise_rmse_used_clean`, `reliability_used_clean`
- gate reason：`layout_gate_reason`, `pointwise_gate_reason`, `reliability_gate_reason`

**实现要点**：
- 缺失率：`df_transparent[col].isna().mean()`
- active_time：missing/==0、p50/p95/p99（必要时 log 统计）
- active_time_source：来源占比（主结论优先以 active_logs 子集；lead_time 作为敏感性验证）
- gate_reason：在 used=False 时 reason 为空/unknown 的比例（防止“静默丢样本”）

````python
# =============================================================================
# [论文表 Table.P0] 字段缺失率与 active_time 覆盖率
# =============================================================================
# 数据集：df_transparent（无过滤）
# 目的：让审稿人明确哪些字段缺失、缺失是否系统性
# =============================================================================

# 定义要统计的列
audit_cols = [
    # 核心字段
    ("scope_clean", "Scope"),
    ("difficulty", "Difficulty"),
    ("model_issue", "Model Issue"),
    ("active_time", "Active Time"),
    # 指标字段
    ("iou", "IoU (Polygon)"),
    ("iou_corner", "IoU (Corner)"),
    ("boundary_rmse_px", "Boundary RMSE"),
    ("layout_2d_iou", "Layout 2D IoU"),
    ("layout_3d_iou", "Layout 3D IoU"),
    # 可靠性字段
    ("iou_to_consensus_loo", "IoU to Consensus (LOO)"),
    ("iou_to_others_median", "IoU to Others Median"),
    # 门控字段
    ("layout_used_clean", "Layout Used"),
    ("pointwise_rmse_used_clean", "Pointwise Used"),
    ("reliability_used_clean", "Reliability Used"),
]

# 计算统计
stats_rows = []
for col, label in audit_cols:
    if col not in df_transparent.columns:
        continue
    
    col_data = df_transparent[col]
    n_total = len(col_data)
    n_missing = col_data.isna().sum()
    missing_rate = n_missing / n_total if n_total > 0 else 0
    
    row = {
        "字段": label,
        "列名": col,
        "总数": n_total,
        "缺失数": n_missing,
        "缺失率": f"{missing_rate:.1%}",
    }
    
    # 对数值列计算额外统计
    if col == "active_time":
        at = to_numeric_safe(col_data)
        n_zero = (at == 0).sum()
        row["零值数"] = n_zero
        row["零值率"] = f"{n_zero / n_total:.1%}" if n_total > 0 else "N/A"
        row["中位数"] = f"{safe_median(at):.1f}s" if safe_median(at) else "N/A"
        row["P95"] = f"{safe_percentile(at, 95):.1f}s" if safe_percentile(at, 95) else "N/A"
    
    stats_rows.append(row)

table_p0 = pd.DataFrame(stats_rows)
display(table_p0)

# 保存表格
save_table(table_p0, "Table_P0_field_audit", category="tables", index=False)

# 额外：active_time 来源分布（如果存在）
if "active_time_source" in df_transparent.columns:
    print("\n📊 Active Time 来源分布:")
    source_dist = df_transparent["active_time_source"].value_counts(dropna=False)
    print(source_dist)
    print(f"\n主结论应优先使用 active_logs 子集 (n={source_dist.get('active_logs', 0)})")
````

## [论文图 Fig.P1 | RQ1] Manual vs Semi 的 active_time 分布（效率主结果）

**放置位置**：主文 RQ1（效率）结果第一张图（最核心）
**回答的问题**：半自动初始化是否降低真实活跃标注时间？降幅是多少？长尾/高工作量子群是否被系统性影响？

**数据集（建议两层口径都算）**：
- overall：`df_transparent`（完整展示耗时，避免只看可算子集）
- in-scope：`df_inscope`（聚焦可复现任务目标）

**最少输出**：
- Panel A：manual vs semi 的 active_time 分布（box/violin + 点）
- Panel B（可选）：按经验/标注者分层（如果 CSV 没有经验字段，就用 annotator 作为 proxy）
- 文本：中位数/IQR；（可选）bootstrap CI

````python
# =============================================================================
# [论文图 Fig.P1 | RQ1] Manual vs Semi 的 active_time 分布（效率主结果）
# =============================================================================
# 数据集：df_inscope（in-scope 子集）
# 目的：半自动初始化是否降低真实活跃标注时间？
# =============================================================================

def plot_efficiency_comparison(
    df: pd.DataFrame,
    time_col: str = "active_time",
    group_col: str = "condition_clean",
    title: str = "Fig.P1: Active Time by Condition (RQ1)",
    log_y: bool = False,
) -> plt.Figure:
    """绘制 Manual vs Semi 的效率对比图。"""
    
    # 数据准备
    df_plot = df[[group_col, time_col]].copy()
    df_plot[time_col] = to_numeric_safe(df_plot[time_col])
    df_plot = df_plot.dropna()
    df_plot["Condition"] = df_plot[group_col].apply(condition_display)
    
    # 计算统计量
    stats = {}
    for cond in ["manual", "semi"]:
        vals = df_plot.loc[df_plot[group_col] == cond, time_col]
        if len(vals) > 0:
            med, ci_lo, ci_hi = bootstrap_ci(vals, np.median)
            stats[cond] = {
                "n": len(vals),
                "median": med,
                "ci_lo": ci_lo,
                "ci_hi": ci_hi,
                "iqr": (vals.quantile(0.25), vals.quantile(0.75)),
            }
    
    # 绘图
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Panel A: Box + Strip plot
    ax1 = axes[0]
    order = ["Manual", "Semi-Auto"]
    sns.boxplot(data=df_plot, x="Condition", y=time_col, order=order, ax=ax1, 
                palette={"Manual": "#4C72B0", "Semi-Auto": "#55A868"}, width=0.5)
    sns.stripplot(data=df_plot, x="Condition", y=time_col, order=order, ax=ax1,
                  color="gray", alpha=0.3, size=3, jitter=0.2)
    ax1.set_xlabel("")
    ax1.set_ylabel("Active Time (seconds)")
    ax1.set_title("Panel A: Distribution Comparison")
    if log_y:
        ax1.set_yscale("log")
    
    # 添加统计注释
    for i, cond in enumerate(["manual", "semi"]):
        if cond in stats:
            s = stats[cond]
            ax1.annotate(
                f"n={s['n']}\nmed={s['median']:.0f}s\n95%CI:[{s['ci_lo']:.0f},{s['ci_hi']:.0f}]",
                xy=(i, s['median']), xytext=(i + 0.3, s['median'] * 1.2),
                fontsize=9, ha="left",
            )
    
    # Panel B: ECDF
    ax2 = axes[1]
    for cond, color in [("manual", "#4C72B0"), ("semi", "#55A868")]:
        vals = df_plot.loc[df_plot[group_col] == cond, time_col].sort_values()
        if len(vals) > 0:
            ecdf = np.arange(1, len(vals) + 1) / len(vals)
            ax2.plot(vals, ecdf, label=condition_display(cond), color=color, linewidth=2)
    ax2.set_xlabel("Active Time (seconds)")
    ax2.set_ylabel("Cumulative Proportion")
    ax2.set_title("Panel B: ECDF")
    ax2.legend()
    ax2.set_xlim(left=0)
    if log_y:
        ax2.set_xscale("log")
    
    fig.suptitle(title, fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    
    return fig, stats

# 绘制并保存
if "df_inscope" in globals():
    print_dataset_stats(df_inscope, "df_inscope (Fig.P1)")
    fig_p1, stats_p1 = plot_efficiency_comparison(df_inscope)
else:
    print_dataset_stats(df_main, "df_main (fallback, Fig.P1)")
    fig_p1, stats_p1 = plot_efficiency_comparison(df_main)

save_figure(fig_p1, "Fig_P1_efficiency_active_time", category="efficiency")
plt.show()

# 打印关键统计
print("\n📊 效率对比统计（RQ1 核心结果）:")
for cond, s in stats_p1.items():
    print(f"  {condition_display(cond)}: n={s['n']}, median={s['median']:.1f}s, 95%CI=[{s['ci_lo']:.1f}, {s['ci_hi']:.1f}]")

if "manual" in stats_p1 and "semi" in stats_p1:
    reduction = (stats_p1["manual"]["median"] - stats_p1["semi"]["median"]) / stats_p1["manual"]["median"]
    print(f"\n  ⏱️  时间节省: {reduction:.1%}")
````

## [补充图 | 数据卫生] active_time 分布与 In-scope 对比

**用途**：数据卫生检查（补充图，不进入 P0-P8/D1-D3 编号体系）。
**数据集**：`df_transparent`
**核心列**：`active_time`, `data_valid_inscope`，（可选）`condition_clean`

**建议输出**：
- active_time 的直方图/ECDF（必要时 log-x）
- In-scope vs 非 In-scope 的分布对比
- 文本注释：active_time==0 比例、p95/p99、长尾样本数量

````python
# =============================================================================
# [补充图] active_time 分布与 In-scope 对比（数据卫生检查）
# =============================================================================

def plot_active_time_hygiene(df: pd.DataFrame) -> plt.Figure:
    """绘制 active_time 的数据卫生检查图。"""
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    
    at = to_numeric_safe(df["active_time"])
    at_valid = at.dropna()
    
    # Panel A: 直方图（log-x）
    ax1 = axes[0]
    at_pos = at_valid[at_valid > 0]
    if len(at_pos) > 0:
        ax1.hist(at_pos, bins=50, edgecolor="white", alpha=0.7)
        ax1.axvline(at_pos.median(), color="red", linestyle="--", label=f"Median: {at_pos.median():.0f}s")
        p95 = np.percentile(at_pos, 95)
        ax1.axvline(p95, color="orange", linestyle=":", label=f"P95: {p95:.0f}s")
    ax1.set_xlabel("Active Time (seconds)")
    ax1.set_ylabel("Count")
    ax1.set_title("Panel A: Distribution (all)")
    ax1.legend(fontsize=9)
    
    # Panel B: In-scope vs 非 In-scope
    ax2 = axes[1]
    for mask, label, color in [
        (df["data_valid_inscope"], "In-scope", "#55A868"),
        (~df["data_valid_inscope"], "Non In-scope", "#C44E52"),
    ]:
        vals = to_numeric_safe(df.loc[mask, "active_time"]).dropna()
        if len(vals) > 0:
            vals_sorted = vals.sort_values()
            ecdf = np.arange(1, len(vals) + 1) / len(vals)
            ax2.plot(vals_sorted, ecdf, label=f"{label} (n={len(vals)})", linewidth=2, color=color)
    ax2.set_xlabel("Active Time (seconds)")
    ax2.set_ylabel("Cumulative Proportion")
    ax2.set_title("Panel B: ECDF by In-scope Status")
    ax2.legend()
    
    # Panel C: 异常统计
    ax3 = axes[2]
    n_total = len(at)
    n_missing = at.isna().sum()
    n_zero = (at == 0).sum()
    n_p99 = (at > np.percentile(at_valid, 99)).sum() if len(at_valid) > 0 else 0
    
    stats_text = f"""Active Time 数据卫生统计
    
总样本数: {n_total}
缺失: {n_missing} ({n_missing/n_total:.1%})
零值: {n_zero} ({n_zero/n_total:.1%})
P99+长尾: {n_p99} ({n_p99/n_total:.1%})

中位数: {safe_median(at):.1f}s
P95: {safe_percentile(at, 95):.1f}s
P99: {safe_percentile(at, 99):.1f}s
"""
    ax3.text(0.1, 0.9, stats_text, transform=ax3.transAxes, fontsize=11,
             verticalalignment="top", fontfamily="monospace",
             bbox=dict(boxstyle="round", facecolor="wheat", alpha=0.5))
    ax3.axis("off")
    ax3.set_title("Panel C: Summary Statistics")
    
    plt.tight_layout()
    return fig

fig_hygiene = plot_active_time_hygiene(df_transparent)
save_figure(fig_hygiene, "Fig_Supp_active_time_hygiene", category="diagnostics")
plt.show()
````

## [论文图 Fig.P2 | 透明披露] Scope / OOS 比例与子类分布

**放置位置**：主文 Section 4（透明披露）或数据集描述处
**回答的问题**：你的任务里有多少比例满足 in-scope 假设？OOS 子类是什么？scope 缺失是否存在？

**数据集**：`df_transparent`
**核心列**：`scope_clean`, `scope_missing`, `is_oos_clean`，（可选）`condition_clean`

**实现建议**：
- 对 `scope_clean.fillna('(missing)')` 做 countplot/barplot（可按 condition 分色）
- 同时输出文本统计：missing 比例、OOS 比例（scope非空条件下）

````python
# =============================================================================
# [论文图 Fig.P2] Scope / OOS 比例与子类分布
# =============================================================================
# 数据集：df_transparent
# 目的：披露任务里有多少满足 in-scope 假设，OOS 子类是什么
# =============================================================================

def plot_scope_distribution(df: pd.DataFrame) -> plt.Figure:
    """绘制 Scope 分布图。"""
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # 准备数据
    scope_data = df["scope_clean"].fillna("(missing)").copy()
    
    # Panel A: 总体 scope 分布
    ax1 = axes[0]
    scope_counts = scope_data.value_counts()
    colors = []
    for s in scope_counts.index:
        if s == "(missing)":
            colors.append("#999999")
        elif is_oos_scope(s):
            colors.append("#C44E52")
        else:
            colors.append("#55A868")
    
    bars = ax1.barh(range(len(scope_counts)), scope_counts.values, color=colors)
    ax1.set_yticks(range(len(scope_counts)))
    ax1.set_yticklabels(scope_counts.index)
    ax1.set_xlabel("Count")
    ax1.set_title("Panel A: Scope Distribution (All)")
    ax1.invert_yaxis()
    
    # 添加数值标签
    for i, (count, pct) in enumerate(zip(scope_counts.values, scope_counts.values / len(df))):
        ax1.text(count + 1, i, f"{count} ({pct:.1%})", va="center", fontsize=9)
    
    # Panel B: 按 condition 分层
    ax2 = axes[1]
    if "condition_clean" in df.columns:
        # 简化为三类：in-scope, OOS, missing
        df_plot = df.copy()
        df_plot["scope_cat"] = df_plot["scope_clean"].apply(
            lambda x: "(missing)" if pd.isna(x) else ("OOS" if is_oos_scope(x) else "In-scope")
        )
        df_plot["Condition"] = df_plot["condition_clean"].apply(condition_display)
        
        ct = pd.crosstab(df_plot["Condition"], df_plot["scope_cat"], normalize="index") * 100
        ct = ct.reindex(columns=["In-scope", "OOS", "(missing)"], fill_value=0)
        
        ct.plot(kind="barh", stacked=True, ax=ax2, 
                color={"In-scope": "#55A868", "OOS": "#C44E52", "(missing)": "#999999"})
        ax2.set_xlabel("Percentage (%)")
        ax2.set_title("Panel B: Scope by Condition")
        ax2.legend(title="Scope", bbox_to_anchor=(1.02, 1), loc="upper left")
    else:
        ax2.text(0.5, 0.5, "condition 字段不存在", ha="center", va="center", transform=ax2.transAxes)
        ax2.axis("off")
    
    plt.tight_layout()
    return fig

print_dataset_stats(df_transparent, "df_transparent (Fig.P2)")
fig_p2 = plot_scope_distribution(df_transparent)
save_figure(fig_p2, "Fig_P2_scope_distribution", category="transparency")
plt.show()

# 打印统计摘要
print("\n📊 Scope 统计摘要:")
print(f"  总样本: {len(df_transparent)}")
print(f"  scope 缺失: {df_transparent['scope_missing'].sum()} ({df_transparent['scope_missing'].mean():.1%})")
valid_scope = df_transparent[~df_transparent["scope_missing"]]
if len(valid_scope) > 0:
    print(f"  OOS 比例（scope非空）: {valid_scope['is_oos_clean'].mean():.1%}")
````

## [补充图 | 透明披露细化] layout 门控失败原因分布（完整展开）

**用途**：当主文 Fig.P3 里只放“Top-K gate reasons”或仅用文字总结时，用这张补充图完整披露。
**数据集**：`df_transparent`（筛选 `layout_used_clean == False`）
**核心列**：`layout_gate_reason`（缺失已补为 `unknown_gate_reason`）

**建议输出**：
- layout_gate_reason 的频数条形图（降序）
- （可选）按 `condition_clean` 分层对比（检查是否存在条件相关的缺失机制）

````python
# =============================================================================
# [补充图] Layout 门控失败原因分布
# =============================================================================
# 数据集：df_transparent（筛选 layout_used_clean == False）
# =============================================================================

def plot_gate_reason_distribution(df: pd.DataFrame) -> plt.Figure:
    """绘制门控失败原因分布。"""
    
    # 筛选被门控的样本
    df_gated = df[df["layout_used_clean"] == False].copy()
    
    if len(df_gated) == 0:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, "No gated samples", ha="center", va="center")
        ax.axis("off")
        return fig
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Panel A: 总体分布
    ax1 = axes[0]
    gate_counts = df_gated["layout_gate_reason"].value_counts()
    
    # Top 10 + Others
    if len(gate_counts) > 10:
        top_10 = gate_counts.head(10)
        others = gate_counts.iloc[10:].sum()
        gate_counts = pd.concat([top_10, pd.Series({"(others)": others})])
    
    colors = ["#C44E52" if "mismatch" in str(r).lower() else "#4C72B0" for r in gate_counts.index]
    ax1.barh(range(len(gate_counts)), gate_counts.values, color=colors)
    ax1.set_yticks(range(len(gate_counts)))
    ax1.set_yticklabels(gate_counts.index)
    ax1.set_xlabel("Count")
    ax1.set_title(f"Panel A: Gate Reasons (n={len(df_gated)} gated)")
    ax1.invert_yaxis()
    
    # 添加百分比标签
    for i, (count, pct) in enumerate(zip(gate_counts.values, gate_counts.values / len(df_gated))):
        ax1.text(count + 0.5, i, f"{pct:.1%}", va="center", fontsize=9)
    
    # Panel B: 按 condition 分层
    ax2 = axes[1]
    if "condition_clean" in df_gated.columns:
        ct = pd.crosstab(df_gated["condition_clean"].apply(condition_display), 
                        df_gated["layout_gate_reason"])
        # 只保留 top 5 reasons
        top_reasons = df_gated["layout_gate_reason"].value_counts().head(5).index.tolist()
        ct = ct.reindex(columns=top_reasons, fill_value=0)
        
        ct.plot(kind="bar", ax=ax2, width=0.7)
        ax2.set_xlabel("")
        ax2.set_ylabel("Count")
        ax2.set_title("Panel B: Gate Reasons by Condition")
        ax2.legend(title="Gate Reason", bbox_to_anchor=(1.02, 1), loc="upper left", fontsize=8)
        ax2.tick_params(axis="x", rotation=0)
    
    plt.tight_layout()
    return fig

fig_gate = plot_gate_reason_distribution(df_transparent)
save_figure(fig_gate, "Fig_Supp_gate_reasons", category="diagnostics")
plt.show()

# P0-1 校验
gate_dist = df_transparent[df_transparent["layout_used_clean"] == False]["layout_gate_reason"].value_counts()
if "n_pairs_mismatch" in gate_dist.index:
    print(f"🔴 REGRESSION: 仍有 {gate_dist['n_pairs_mismatch']} 个样本因 n_pairs_mismatch 被门控")
else:
    print("✅ P0-1 校验通过：layout_gate_reason 中未出现 n_pairs_mismatch")
````

## [论文图 Fig.P4 | RQ1/RQ2 背景] 改动幅度/结构差异分布（工作量子群）

**放置位置**：主文 RQ1 背景或透明披露后（用于解释工作量异质性）
**回答的问题**：数据里是否存在“强纠错/高工作量子群”？它是否与 mismatch（例如 |Δn_pairs|>0 或 n_pairs_mismatch）强相关？

**数据集**：优先 `df_inscope`（覆盖强纠错样本），必要时对比 `df_main`（标准指标可算子集）

**建议输出**：
- Panel A：`active_time` 的分位数/ECDF（可附一句长尾统计）
- Panel B：`boundary_rmse_px` 分布（.notna() 后）
- Panel C：`abs_delta_n_pairs`（或等价列）分布 + 分箱计数
- Panel D（可选）：按 mismatch vs 非 mismatch 分层的对比（与 Fig.P3 的 Panel C 呼应）

````python
# =============================================================================
# [论文图 Fig.P4] 改动幅度/结构差异分布（工作量子群）
# =============================================================================
# 数据集：df_inscope
# 目的：展示"强纠错/高工作量子群"的存在
# =============================================================================

def plot_edit_magnitude_distribution(df: pd.DataFrame) -> plt.Figure:
    """绘制改动幅度分布图。"""
    
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    # Panel A: |Δn_pairs| 分布
    ax1 = axes[0, 0]
    if all(c in df.columns for c in ["pred_n_pairs", "ann_n_pairs"]):
        delta = (to_numeric_safe(df["pred_n_pairs"]) - to_numeric_safe(df["ann_n_pairs"])).abs()
        delta_valid = delta.dropna()
        
        counts = delta_valid.value_counts().sort_index()
        ax1.bar(counts.index[:10], counts.values[:10], color="#4C72B0", edgecolor="white")
        ax1.set_xlabel("|Δn_pairs|")
        ax1.set_ylabel("Count")
        ax1.set_title(f"Panel A: |Δn_pairs| Distribution (n={len(delta_valid)})")
        
        # 标注 mismatch 比例
        mismatch_rate = (delta_valid > 0).mean()
        ax1.annotate(f"Mismatch rate: {mismatch_rate:.1%}", xy=(0.95, 0.95), 
                    xycoords="axes fraction", ha="right", va="top", fontsize=10,
                    bbox=dict(boxstyle="round", facecolor="wheat"))
    else:
        ax1.text(0.5, 0.5, "pred_n_pairs/ann_n_pairs 不存在", ha="center", va="center")
        ax1.axis("off")
    
    # Panel B: Boundary RMSE 分布
    ax2 = axes[0, 1]
    if "boundary_rmse_px" in df.columns:
        rmse = to_numeric_safe(df["boundary_rmse_px"]).dropna()
        if len(rmse) > 0:
            ax2.hist(rmse, bins=50, color="#55A868", edgecolor="white", alpha=0.7)
            ax2.axvline(rmse.median(), color="red", linestyle="--", label=f"Median: {rmse.median():.1f}px")
            ax2.set_xlabel("Boundary RMSE (pixels)")
            ax2.set_ylabel("Count")
            ax2.set_title(f"Panel B: Boundary RMSE Distribution (n={len(rmse)})")
            ax2.legend()
    else:
        ax2.text(0.5, 0.5, "boundary_rmse_px 不存在", ha="center", va="center")
        ax2.axis("off")
    
    # Panel C: IoU 分布
    ax3 = axes[1, 0]
    iou_col = "iou" if "iou" in df.columns else "iou_corner"
    if iou_col in df.columns:
        iou = to_numeric_safe(df[iou_col]).dropna()
        if len(iou) > 0:
            ax3.hist(iou, bins=50, color="#C44E52", edgecolor="white", alpha=0.7)
            ax3.axvline(iou.median(), color="blue", linestyle="--", label=f"Median: {iou.median():.3f}")
            ax3.set_xlabel(f"{iou_col}")
            ax3.set_ylabel("Count")
            ax3.set_title(f"Panel C: {iou_col} Distribution (n={len(iou)})")
            ax3.legend()
    else:
        ax3.text(0.5, 0.5, "iou 字段不存在", ha="center", va="center")
        ax3.axis("off")
    
    # Panel D: Mismatch vs Matched 的 active_time 对比
    ax4 = axes[1, 1]
    if all(c in df.columns for c in ["pred_n_pairs", "ann_n_pairs", "active_time"]):
        delta = to_numeric_safe(df["pred_n_pairs"]) - to_numeric_safe(df["ann_n_pairs"])
        df_temp = df.copy()
        df_temp["_mismatch"] = (delta.notna() & delta.ne(0)).map({True: "Mismatch", False: "Matched"})
        df_temp["_active_time"] = to_numeric_safe(df_temp["active_time"])
        df_temp = df_temp.dropna(subset=["_mismatch", "_active_time"])
        
        if len(df_temp) > 0:
            sns.boxplot(data=df_temp, x="_mismatch", y="_active_time", ax=ax4,
                       order=["Matched", "Mismatch"], palette=["#55A868", "#C44E52"])
            ax4.set_xlabel("")
            ax4.set_ylabel("Active Time (seconds)")
            ax4.set_title("Panel D: Active Time by Mismatch Status")
            
            # 添加统计
            for i, status in enumerate(["Matched", "Mismatch"]):
                vals = df_temp.loc[df_temp["_mismatch"] == status, "_active_time"]
                if len(vals) > 0:
                    med = vals.median()
                    ax4.annotate(f"n={len(vals)}\nmed={med:.0f}s", 
                               xy=(i, med), xytext=(i + 0.25, med),
                               fontsize=9, ha="left")
    else:
        ax4.text(0.5, 0.5, "缺少必要字段", ha="center", va="center")
        ax4.axis("off")
    
    plt.suptitle("Fig.P4: Edit Magnitude & Workload Subgroups", fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    return fig

fig_p4 = plot_edit_magnitude_distribution(df_inscope)
save_figure(fig_p4, "Fig_P4_edit_magnitude", category="efficiency")
plt.show()
````

## [论文图 Fig.P5 | RQ1] active_time vs 改动幅度/质量 proxy（log-x）

**放置位置**：主文 RQ1（效率）结果段落
**回答的问题**：省时是否来自“少改/瞎改”？耗时与改动幅度/质量 proxy 的关系是否一致？

**数据集（建议两条线并列）**：
- 透明对比：`df_inscope`（覆盖强纠错样本，使用结构无关 proxy）
- 主文口径：`df_main`（当你需要报告标准 layout 指标时）

**核心列（按你最终选定指标）**：
- x：`active_time`（log）
- y（至少选其一）：`iou`/`iou_corner`（可用时）、`boundary_rmse_px`、`iou_to_others_median`（若用作质量 proxy）
- 分层：`condition_clean`（manual/semi）
- 分层（可选）：`abs_delta_n_pairs`（或等价列，做分层/分面）

**统计建议**：Spearman 相关 + 分层后再算（避免 Simpson’s paradox）。

````python
# =============================================================================
# [论文图 Fig.P5] active_time vs 改动幅度/质量 proxy（log-x）
# =============================================================================
# 数据集：df_inscope / df_main
# 目的：省时是否来自"少改/瞎改"？
# =============================================================================

def plot_time_vs_quality(df: pd.DataFrame, title_suffix: str = "") -> plt.Figure:
    """绘制耗时与质量/改动幅度的关系图。"""
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # 准备数据
    df_plot = df.copy()
    df_plot["active_time"] = to_numeric_safe(df_plot["active_time"])
    df_plot["Condition"] = df_plot["condition_clean"].apply(condition_display)
    
    # Panel A: active_time vs boundary_rmse
    ax1 = axes[0]
    if "boundary_rmse_px" in df_plot.columns:
        df_plot["boundary_rmse_px"] = to_numeric_safe(df_plot["boundary_rmse_px"])
        df_sub = df_plot.dropna(subset=["active_time", "boundary_rmse_px"])
        
        for cond, color in [("Manual", "#4C72B0"), ("Semi-Auto", "#55A868")]:
            sub = df_sub[df_sub["Condition"] == cond]
            if len(sub) > 0:
                ax1.scatter(sub["active_time"], sub["boundary_rmse_px"], 
                          alpha=0.5, label=cond, color=color, s=20)
        
        ax1.set_xlabel("Active Time (seconds, log)")
        ax1.set_ylabel("Boundary RMSE (pixels)")
        ax1.set_title("Panel A: Time vs Boundary RMSE")
        ax1.set_xscale("log")
        ax1.legend()
        
        # Spearman 相关
        from scipy.stats import spearmanr
        r, p = spearmanr(df_sub["active_time"], df_sub["boundary_rmse_px"])
        ax1.annotate(f"Spearman r={r:.3f}\np={p:.2e}", xy=(0.05, 0.95),
                    xycoords="axes fraction", va="top", fontsize=9,
                    bbox=dict(boxstyle="round", facecolor="wheat"))
    else:
        ax1.text(0.5, 0.5, "boundary_rmse_px 不存在", ha="center", va="center")
        ax1.axis("off")
    
    # Panel B: active_time vs iou
    ax2 = axes[1]
    iou_col = "iou" if "iou" in df_plot.columns else "iou_corner"
    if iou_col in df_plot.columns:
        df_plot[iou_col] = to_numeric_safe(df_plot[iou_col])
        df_sub = df_plot.dropna(subset=["active_time", iou_col])
        
        for cond, color in [("Manual", "#4C72B0"), ("Semi-Auto", "#55A868")]:
            sub = df_sub[df_sub["Condition"] == cond]
            if len(sub) > 0:
                ax2.scatter(sub["active_time"], sub[iou_col], 
                          alpha=0.5, label=cond, color=color, s=20)
        
        ax2.set_xlabel("Active Time (seconds, log)")
        ax2.set_ylabel(f"{iou_col}")
        ax2.set_title(f"Panel B: Time vs {iou_col}")
        ax2.set_xscale("log")
        ax2.legend()
        
        r, p = spearmanr(df_sub["active_time"], df_sub[iou_col])
        ax2.annotate(f"Spearman r={r:.3f}\np={p:.2e}", xy=(0.05, 0.05),
                    xycoords="axes fraction", va="bottom", fontsize=9,
                    bbox=dict(boxstyle="round", facecolor="wheat"))
    else:
        ax2.text(0.5, 0.5, "iou 字段不存在", ha="center", va="center")
        ax2.axis("off")
    
    # Panel C: active_time vs iou_to_others_median（质量 proxy）
    ax3 = axes[2]
    if "iou_to_others_median" in df_plot.columns:
        df_plot["iou_to_others_median"] = to_numeric_safe(df_plot["iou_to_others_median"])
        df_sub = df_plot.dropna(subset=["active_time", "iou_to_others_median"])
        
        for cond, color in [("Manual", "#4C72B0"), ("Semi-Auto", "#55A868")]:
            sub = df_sub[df_sub["Condition"] == cond]
            if len(sub) > 0:
                ax3.scatter(sub["active_time"], sub["iou_to_others_median"], 
                          alpha=0.5, label=cond, color=color, s=20)
        
        ax3.set_xlabel("Active Time (seconds, log)")
        ax3.set_ylabel("IoU to Others Median (Agreement)")
        ax3.set_title("Panel C: Time vs Agreement")
        ax3.set_xscale("log")
        ax3.legend()
        
        r, p = spearmanr(df_sub["active_time"], df_sub["iou_to_others_median"])
        ax3.annotate(f"Spearman r={r:.3f}\np={p:.2e}", xy=(0.05, 0.05),
                    xycoords="axes fraction", va="bottom", fontsize=9,
                    bbox=dict(boxstyle="round", facecolor="wheat"))
    else:
        ax3.text(0.5, 0.5, "iou_to_others_median 不存在", ha="center", va="center")
        ax3.axis("off")
    
    plt.suptitle(f"Fig.P5: Active Time vs Quality/Edit Metrics{title_suffix}", 
                fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    return fig

# 分别对 df_inscope 和 df_main 绘图
fig_p5_inscope = plot_time_vs_quality(df_inscope, " (In-scope)")
save_figure(fig_p5_inscope, "Fig_P5_time_vs_quality_inscope", category="efficiency")
plt.show()

if len(df_main) != len(df_inscope):
    fig_p5_main = plot_time_vs_quality(df_main, " (Main)")
    save_figure(fig_p5_main, "Fig_P5_time_vs_quality_main", category="efficiency")
    plt.show()
````

## [补充图 | 标注者差异] 标注者间主指标分布（按标注者）

**用途**：主要用于解释个体差异、支撑 RQ3 的叙述（补充图）。
**数据集**：建议 `df_main`（主指标口径一致），也可对比 `df_transparent`
**核心列**：`annotator_id` + 主指标列（如 `iou`/`iou_corner`）

**建议输出**：boxplot + stripplot，并给出每个标注者的 n（样本量）。

````python
# =============================================================================
# [补充图] 标注者间主指标分布（按标注者）
# =============================================================================

def plot_annotator_comparison(df: pd.DataFrame, metric_col: str = "iou") -> plt.Figure:
    """绘制按标注者分层的指标分布。"""
    
    if metric_col not in df.columns:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, f"{metric_col} 字段不存在", ha="center", va="center")
        ax.axis("off")
        return fig
    
    df_plot = df.copy()
    df_plot[metric_col] = to_numeric_safe(df_plot[metric_col])
    df_plot = df_plot.dropna(subset=[metric_col])
    
    # 按标注者统计样本量并排序
    annotator_stats = df_plot.groupby("annotator_id").agg({
        metric_col: ["count", "median"]
    }).droplevel(0, axis=1)
    annotator_stats.columns = ["n", "median"]
    annotator_stats = annotator_stats.sort_values("median", ascending=False)
    
    # 只显示 top 20 标注者
    top_annotators = annotator_stats.head(20).index.tolist()
    df_plot = df_plot[df_plot["annotator_id"].isin(top_annotators)]
    
    fig, ax = plt.subplots(figsize=(14, 6))
    
    sns.boxplot(data=df_plot, x="annotator_id", y=metric_col, 
               order=top_annotators, ax=ax, palette="viridis")
    sns.stripplot(data=df_plot, x="annotator_id", y=metric_col,
                 order=top_annotators, ax=ax, color="gray", alpha=0.3, size=2)
    
    ax.set_xlabel("Annotator ID")
    ax.set_ylabel(metric_col)
    ax.set_title(f"Annotator Performance: {metric_col} (Top 20 by median)")
    ax.tick_params(axis="x", rotation=45)
    
    # 添加样本量标签
    for i, uid in enumerate(top_annotators):
        n = annotator_stats.loc[uid, "n"]
        ax.annotate(f"n={n}", xy=(i, ax.get_ylim()[0]), fontsize=7, ha="center", va="top")
    
    plt.tight_layout()
    return fig

iou_col = "iou" if "iou" in df_main.columns else "iou_corner"
fig_annotator = plot_annotator_comparison(df_main, iou_col)
save_figure(fig_annotator, "Fig_Supp_annotator_comparison", category="diagnostics")
plt.show()
````

## [论文图 Fig.P7 | RQ2/RQ3] 标注者一致性对比——个体 vs 共识 / 个体 vs 同行（可靠性集）

**数据集**：`df_rel`（多标注 + LOO 指标可用 + `reliability_used=True`）
**核心列**：
- `iou_to_consensus_loo`：个体与 leave-one-out 共识的 IoU
- `iou_to_others_median`：个体与其他标注者的中位 IoU
- （可选）`condition_clean`, `task_scope_n_total`

**目的**：论文"可靠性/一致性"关键证据：说明标注者对同一任务的收敛程度，以及共识稳定性。

**若有 Calibration_semi**：
- 应分别列出 Manual vs Semi 的一致性分布对比
- 图注中明确标注数据来源集（`C_manual` 与 `C_semi` 的重复标注）
- 展示 manual 与 semi 条件下标注者收敛程度的差异

**注意事项**：
- n=2 任务的 LOO 退化为 pairwise，需在论文方法部分明确；可在图中用分层/标注区分 n=2 vs n≥3
- `df_rel` 的过滤口径必须与 Table.P0 的门控披露一致

**你要写的代码**：
1. 分布并列（box/violin）或散点对比（x=others_median, y=consensus_loo）
2. （可选）按 `condition_clean` 或 `task_scope_n_total` 分层
3. 输出关键统计：中位数、IQR、低分位数

````python
# =============================================================================
# [论文图 Fig.P7] 标注者一致性对比——个体 vs 共识 / 个体 vs 同行
# =============================================================================
# 数据集：df_rel_manual / df_rel_semi（可靠性集）
# 目的：展示标注者对同一任务的收敛程度
# =============================================================================

def plot_reliability_comparison(df: pd.DataFrame) -> plt.Figure:
    """绘制可靠性指标对比图。"""
    
    required_cols = ["iou_to_consensus_loo", "iou_to_others_median"]
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, f"缺少字段: {missing}", ha="center", va="center")
        ax.axis("off")
        return fig
    
    df_plot = df.copy()
    for col in required_cols:
        df_plot[col] = to_numeric_safe(df_plot[col])
    df_plot = df_plot.dropna(subset=required_cols)
    
    if "condition_clean" in df_plot.columns:
        df_plot["Condition"] = df_plot["condition_clean"].apply(condition_display)
    else:
        df_plot["Condition"] = "Unknown"
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Panel A: 并列 boxplot
    ax1 = axes[0]
    df_melt = df_plot.melt(
        id_vars=["Condition"],
        value_vars=required_cols,
        var_name="Metric",
        value_name="IoU"
    )
    df_melt["Metric"] = df_melt["Metric"].map({
        "iou_to_consensus_loo": "To LOO Consensus",
        "iou_to_others_median": "To Others Median"
    })
    
    sns.boxplot(data=df_melt, x="Metric", y="IoU", hue="Condition", ax=ax1)
    ax1.set_xlabel("")
    ax1.set_ylabel("IoU")
    ax1.set_title("Panel A: Reliability Metrics by Condition")
    ax1.legend(title="Condition")
    
    # Panel B: 散点对比
    ax2 = axes[1]
    for cond, color in [("Manual", "#4C72B0"), ("Semi-Auto", "#55A868")]:
        sub = df_plot[df_plot["Condition"] == cond]
        if len(sub) > 0:
            ax2.scatter(sub["iou_to_others_median"], sub["iou_to_consensus_loo"],
                       alpha=0.5, label=f"{cond} (n={len(sub)})", color=color, s=20)
    
    # 对角线
    lims = [0, 1]
    ax2.plot(lims, lims, "k--", alpha=0.5, label="x=y")
    ax2.set_xlabel("IoU to Others Median")
    ax2.set_ylabel("IoU to LOO Consensus")
    ax2.set_title("Panel B: Metric Comparison")
    ax2.set_xlim(lims)
    ax2.set_ylim(lims)
    ax2.legend(fontsize=9)
    ax2.set_aspect("equal")
    
    # Panel C: 分布统计
    ax3 = axes[2]
    stats_text = "📊 Reliability Statistics\n" + "=" * 30 + "\n\n"
    
    for cond in ["manual", "semi"]:
        cond_display = condition_display(cond)
        sub = df_plot[df_plot["condition_clean"] == cond] if "condition_clean" in df_plot.columns else df_plot
        if len(sub) == 0:
            continue
        
        loo_med, loo_lo, loo_hi = bootstrap_ci(sub["iou_to_consensus_loo"])
        others_med, others_lo, others_hi = bootstrap_ci(sub["iou_to_others_median"])
        
        stats_text += f"{cond_display} (n={len(sub)}):\n"
        if loo_med:
            stats_text += f"  LOO Consensus: {loo_med:.3f} [{loo_lo:.3f}, {loo_hi:.3f}]\n"
        if others_med:
            stats_text += f"  Others Median: {others_med:.3f} [{others_lo:.3f}, {others_hi:.3f}]\n"
        stats_text += "\n"
    
    ax3.text(0.05, 0.95, stats_text, transform=ax3.transAxes, fontsize=10,
            verticalalignment="top", fontfamily="monospace",
            bbox=dict(boxstyle="round", facecolor="wheat", alpha=0.5))
    ax3.axis("off")
    ax3.set_title("Panel C: Summary")
    
    plt.suptitle("Fig.P7: Annotator Agreement Analysis (RQ2/RQ3)", fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    return fig

# RQ2 对照：Calibration_manual vs Calibration_semi
print_dataset_stats(df_rel_manual, "df_rel_manual (Fig.P7)")
print_dataset_stats(df_rel_semi, "df_rel_semi (Fig.P7)")

if len(df_rel_manual) + len(df_rel_semi) > 0:
    df_rel_rq2 = pd.concat([df_rel_manual, df_rel_semi], ignore_index=True)
    if "dataset_group" in df_rel_rq2.columns:
        df_rel_rq2["condition_clean"] = df_rel_rq2["dataset_group"].map({
            "Calibration_manual": "manual",
            "Calibration_semi": "semi",
        }).fillna(df_rel_rq2.get("condition_clean", "unknown"))
    fig_p7 = plot_reliability_comparison(df_rel_rq2)
    save_figure(fig_p7, "Fig_P7_reliability_comparison", category="reliability")
    plt.show()
else:
    print("⚠️  df_rel_manual / df_rel_semi 为空，跳过 Fig.P7。请确认已导入 Calibration_* 且 analysis_role=reliability 的数据。")
````

## [论文图 Fig.P8 | RQ2] agreement/IAA 分布（长尾）与异常低协议样本提示（可靠性集）

**数据集**：`df_rel`（已包含 `reliability_used=True` 过滤）
**核心列**：

**目的**：给出 IAA 的整体分布与长尾，支撑论文对标注一致性与可靠性结论，并为"反例/任务重分配"提供可操作阈值（审计用途）。

**若有 Calibration_semi**：

- 图注中标明来源集（重复标注数据）
- 可对比两种条件下的长尾分布差异，为方法论提供证据

**建议输出**：
- 分布图（hist/ECDF/violin）展示 `iou_to_others_median`
- 标注长尾：IQR 下界/低分位数，并输出对应 task 清单（可与 Table.A1 联动）
- （可选）分层报告：n≥3 vs n=2 任务的 IAA 差异

````python
# =============================================================================
# [论文图 Fig.P8] IAA 分布（长尾）与异常低协议样本
# =============================================================================
# 数据集：df_rel_manual / df_rel_semi
# 目的：展示 IAA 的整体分布与长尾
# =============================================================================

def plot_iaa_distribution(df: pd.DataFrame) -> plt.Figure:
    """绘制 IAA 分布与长尾分析。"""
    
    if "iou_to_others_median" not in df.columns:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, "iou_to_others_median 不存在", ha="center", va="center")
        ax.axis("off")
        return fig
    
    df_plot = df.copy()
    df_plot["iou_to_others_median"] = to_numeric_safe(df_plot["iou_to_others_median"])
    df_plot = df_plot.dropna(subset=["iou_to_others_median"])
    
    if "condition_clean" in df_plot.columns:
        df_plot["Condition"] = df_plot["condition_clean"].apply(condition_display)
    else:
        df_plot["Condition"] = "All"
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Panel A: 直方图
    ax1 = axes[0]
    for cond, color in [("Manual", "#4C72B0"), ("Semi-Auto", "#55A868")]:
        sub = df_plot[df_plot["Condition"] == cond]["iou_to_others_median"]
        if len(sub) > 0:
            ax1.hist(sub, bins=30, alpha=0.6, label=f"{cond} (n={len(sub)})", 
                    color=color, edgecolor="white")
    
    ax1.axvline(df_plot["iou_to_others_median"].median(), color="red", linestyle="--", 
               label=f"Overall Median: {df_plot['iou_to_others_median'].median():.3f}")
    ax1.set_xlabel("IoU to Others Median (Agreement)")
    ax1.set_ylabel("Count")
    ax1.set_title("Panel A: IAA Distribution")
    ax1.legend(fontsize=9)
    
    # Panel B: ECDF + 分位数标注
    ax2 = axes[1]
    vals = df_plot["iou_to_others_median"].sort_values()
    ecdf = np.arange(1, len(vals) + 1) / len(vals)
    ax2.plot(vals, ecdf, color="#4C72B0", linewidth=2)
    
    # 标注关键分位数
    for q, style in [(10, ":"), (25, "--")]:
        pct = np.percentile(vals, q)
        ax2.axvline(pct, color="red", linestyle=style, alpha=0.7, label=f"P{q}: {pct:.3f}")
        ax2.axhline(q / 100, color="gray", linestyle=style, alpha=0.3)
    
    ax2.set_xlabel("IoU to Others Median")
    ax2.set_ylabel("Cumulative Proportion")
    ax2.set_title("Panel B: ECDF with Quantiles")
    ax2.legend(fontsize=9)
    ax2.set_xlim(0, 1)
    
    # Panel C: 长尾样本清单
    ax3 = axes[2]
    p10 = np.percentile(df_plot["iou_to_others_median"], 10)
    low_agreement = df_plot[df_plot["iou_to_others_median"] < p10]
    
    stats_text = f"📋 Low Agreement Samples (< P10 = {p10:.3f})\n"
    stats_text += "=" * 40 + "\n\n"
    stats_text += f"总计: {len(low_agreement)} 个样本\n\n"
    
    if "task_id" in low_agreement.columns:
        task_counts = low_agreement["task_id"].value_counts().head(5)
        stats_text += "Top 5 Tasks with Low Agreement:\n"
        for tid, cnt in task_counts.items():
            stats_text += f"  {tid}: {cnt} annotations\n"
    
    ax3.text(0.05, 0.95, stats_text, transform=ax3.transAxes, fontsize=10,
            verticalalignment="top", fontfamily="monospace",
            bbox=dict(boxstyle="round", facecolor="lightyellow", alpha=0.8))
    ax3.axis("off")
    ax3.set_title("Panel C: Low Agreement Audit")
    
    plt.suptitle("Fig.P8: Inter-Annotator Agreement Distribution (RQ2)", fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    return fig

# RQ2 对照：Calibration_manual vs Calibration_semi
print_dataset_stats(df_rel_manual, "df_rel_manual (Fig.P8)")
print_dataset_stats(df_rel_semi, "df_rel_semi (Fig.P8)")

if len(df_rel_manual) + len(df_rel_semi) > 0:
    df_rel_rq2 = pd.concat([df_rel_manual, df_rel_semi], ignore_index=True)
    if "dataset_group" in df_rel_rq2.columns:
        df_rel_rq2["condition_clean"] = df_rel_rq2["dataset_group"].map({
            "Calibration_manual": "manual",
            "Calibration_semi": "semi",
        }).fillna(df_rel_rq2.get("condition_clean", "unknown"))
    fig_p8 = plot_iaa_distribution(df_rel_rq2)
    save_figure(fig_p8, "Fig_P8_iaa_distribution", category="reliability")
    plt.show()
else:
    print("⚠️  df_rel_manual / df_rel_semi 为空，跳过 Fig.P8。请确认已导入 Calibration_* 且 analysis_role=reliability 的数据。")
````

# Section 5: RQ3 - Expert Identification & Task Assignment

- Fig.P6：标注者可靠度 $r_u$ + CI
- （可选）与耗时/任务分配策略关系

## [论文图 Fig.P6 | RQ3] 标注者可靠度 $r_u$ + Bootstrap CI（可靠性集）

**放置位置**：主文 RQ3（专家识别/任务分配）
**回答的问题**：能否用可靠性指标区分标注者能力，并为任务分配提供可解释依据？

**数据集**：`df_rel`（多标注 + 共识/同行指标可用）

**定义建议（可审计、可复现）**：
- 对每个标注者 $u$：$r_u = \mathrm{median}(\mathrm{IoU}_{u,\text{consensus-LOO}})$ 或 $\mathrm{median}(\mathrm{IoU}_{u,\text{others-median}})$
- 置信区间：对标注者内的任务行做 bootstrap（重复采样任务行）得到 95% CI

**最少输出**：
- Panel A：每个标注者的 $r_u$ 点估计 + 95% CI（带 n=任务数）
- Panel B（可选）：$r_u$ 与该标注者的耗时中位数关系（为“分配策略”提供证据）

````python
# =============================================================================
# [论文图 Fig.P6] 标注者可靠度 r_u + Bootstrap CI
# =============================================================================
# 数据集：df_rel_manual（仅 Calibration_manual）
# 目的：展示标注者能力区分度
# =============================================================================

def plot_annotator_reliability(df: pd.DataFrame, metric_col: str = "iou_to_consensus_loo") -> plt.Figure:
    """绘制标注者可靠度排名图。"""
    
    if metric_col not in df.columns:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, f"{metric_col} 不存在", ha="center", va="center")
        ax.axis("off")
        return fig
    
    df_plot = df.copy()
    df_plot[metric_col] = to_numeric_safe(df_plot[metric_col])
    df_plot = df_plot.dropna(subset=[metric_col, "annotator_id"])
    
    # 按标注者计算 r_u 及其 bootstrap CI
    annotator_stats = []
    for uid in df_plot["annotator_id"].unique():
        vals = df_plot.loc[df_plot["annotator_id"] == uid, metric_col]
        if len(vals) < 3:  # 至少3个任务才报告 CI
            continue
        
        r_u, ci_lo, ci_hi = bootstrap_ci(vals, np.median)
        annotator_stats.append({
            "annotator_id": uid,
            "r_u": r_u,
            "ci_lo": ci_lo,
            "ci_hi": ci_hi,
            "n_tasks": len(vals),
        })
    
    if not annotator_stats:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, "没有足够数据计算 r_u", ha="center", va="center")
        ax.axis("off")
        return fig
    
    stats_df = pd.DataFrame(annotator_stats).sort_values("r_u", ascending=False)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, max(6, len(stats_df) * 0.3)))
    
    # Panel A: r_u 排名条形图 + CI
    ax1 = axes[0]
    y_pos = range(len(stats_df))
    
    # 绘制 CI 误差条
    ax1.barh(y_pos, stats_df["r_u"], xerr=[
        stats_df["r_u"] - stats_df["ci_lo"],
        stats_df["ci_hi"] - stats_df["r_u"]
    ], color="#4C72B0", alpha=0.7, capsize=3)
    
    ax1.set_yticks(y_pos)
    ax1.set_yticklabels([f"{uid} (n={n})" for uid, n in zip(stats_df["annotator_id"], stats_df["n_tasks"])])
    ax1.set_xlabel(f"$r_u$ = median({metric_col})")
    ax1.set_title("Panel A: Annotator Reliability Ranking")
    ax1.set_xlim(0, 1)
    ax1.invert_yaxis()
    
    # 添加数值标签
    for i, row in enumerate(stats_df.itertuples()):
        ax1.text(row.r_u + 0.02, i, f"{row.r_u:.3f}", va="center", fontsize=8)
    
    # Panel B: r_u vs 任务数散点图
    ax2 = axes[1]
    ax2.scatter(stats_df["n_tasks"], stats_df["r_u"], s=60, alpha=0.7, c="#55A868")
    ax2.set_xlabel("Number of Tasks")
    ax2.set_ylabel("$r_u$")
    ax2.set_title("Panel B: Reliability vs Task Count")
    ax2.set_ylim(0, 1)
    
    # 添加标注者 ID 标签
    for _, row in stats_df.iterrows():
        ax2.annotate(row["annotator_id"], (row["n_tasks"], row["r_u"]),
                    fontsize=7, alpha=0.7, xytext=(3, 3), textcoords="offset points")
    
    plt.suptitle("Fig.P6: Annotator Reliability $r_u$ with Bootstrap CI (RQ3)", 
                fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    return fig, stats_df

print_dataset_stats(df_rel_manual, "df_rel_manual (Fig.P6)")
if len(df_rel_manual) > 0:
    fig_p6, reliability_df = plot_annotator_reliability(df_rel_manual)
    save_figure(fig_p6, "Fig_P6_annotator_reliability", category="reliability")
    save_table(reliability_df, "Table_annotator_reliability", category="tables", index=False)
    plt.show()
else:
    print("⚠️  df_rel_manual 为空，跳过 Fig.P6。请确认已导入 Calibration_manual 且 analysis_role=reliability 的数据。")
````

# Section 6: Sensitivity Analysis (P0-1 Comparison)

- 旧 gate（含 n_pairs_mismatch） vs P0-1
- 覆盖率与主结论稳定性对比

## [敏感性分析 | P0-1] 旧 gate vs P0-1 对比（覆盖率与结论稳定性）

**目的**：验证移除 `n_pairs_mismatch` gate 后，覆盖率显著提升且 RQ1/RQ2 结论保持稳定。

**最少输出**：
- 覆盖率对比（旧口径 vs P0-1）
- RQ1：Manual vs Semi 的中位耗时差是否稳定
- RQ2：IAA/一致性分布的中位数差异是否稳定
- mismatch 子群与 matched 子群的工作量差异（例如中位耗时）

````python
# =============================================================================
# [敏感性分析] 旧 gate vs P0-1 对比表
# =============================================================================
# 目的：验证移除 n_pairs_mismatch gate 后，覆盖率显著提升且 RQ1/RQ2 结论保持稳定。
# =============================================================================

# 在 df_clean 上模拟旧逻辑（包含 n_pairs_mismatch gate），对比覆盖率与结论稳定性
df_old_logic = df_clean.copy()
if all(c in df_old_logic.columns for c in ["layout_used_clean", "pred_n_pairs", "ann_n_pairs"]):
    df_old_logic["layout_used_OLD"] = df_old_logic["layout_used_clean"] & (
        df_old_logic["pred_n_pairs"] == df_old_logic["ann_n_pairs"]
    )
else:
    df_old_logic["layout_used_OLD"] = df_old_logic.get("layout_used_clean", pd.Series([pd.NA] * len(df_old_logic))).fillna(False)

# 1) 覆盖率对比（旧 vs 新）
# BUG-FIX: 应从 df_old_logic 筛选，而不是 df_clean（否则没有 layout_used_OLD 列）
df_sensitivity = df_old_logic[df_old_logic["data_valid_inscope"]].copy()
coverage_new = float(df_sensitivity["layout_used_clean"].mean())
coverage_old = float(df_sensitivity["layout_used_OLD"].mean())
print(f"Coverage NEW (P0-1): {coverage_new:.1%}")
print(f"Coverage OLD (gate mismatch): {coverage_old:.1%}")

# 2) RQ1 稳定性（manual vs semi）
rq1_rows = []
if "condition_clean" in df_sensitivity.columns and "active_time" in df_sensitivity.columns:
    for label, mask in [("NEW", df_sensitivity["layout_used_clean"].eq(True)), ("OLD", df_sensitivity["layout_used_OLD"].eq(True))]:
        sub = df_sensitivity[mask]
        for cond in ["manual", "semi"]:
            v = sub.loc[sub["condition_clean"] == cond, "active_time"].dropna()
            if len(v) > 0:
                med = float(v.median())
                rq1_rows.append({"scope": label, "condition": cond, "active_time_med": med})
                print(f"{label} {cond} active_time median: {med:.1f}s")

# 3) RQ2 稳定性（IAA proxy）
rq2_rows = []
if "iou_to_others_median" in df_sensitivity.columns:
    for label, mask in [("NEW", df_sensitivity["layout_used_clean"].eq(True)), ("OLD", df_sensitivity["layout_used_OLD"].eq(True))]:
        v = df_sensitivity.loc[mask, "iou_to_others_median"].dropna()
        if len(v) > 0:
            med = float(v.median())
            rq2_rows.append({"scope": label, "IAA_med": med})
            print(f"{label} IAA median: {med:.3f}")

# 4) mismatch 子群 vs matched 子群（工作量证据）
# P0-1 后 mismatch 不再是 layout_gate_reason，而是结构差异（pred_n_pairs != ann_n_pairs）
mismatch_summary = {}
if all(c in df_sensitivity.columns for c in ["pred_n_pairs", "ann_n_pairs", "active_time"]):
    pred_pairs = pd.to_numeric(df_sensitivity["pred_n_pairs"], errors="coerce")
    ann_pairs = pd.to_numeric(df_sensitivity["ann_n_pairs"], errors="coerce")
    mismatch_mask = pred_pairs.notna() & ann_pairs.notna() & pred_pairs.ne(ann_pairs)
    matched_mask = pred_pairs.notna() & ann_pairs.notna() & pred_pairs.eq(ann_pairs)
    df_mismatch = df_sensitivity[mismatch_mask]
    df_matched = df_sensitivity[matched_mask]
    if len(df_mismatch) > 0 and len(df_matched) > 0:
        m_med = float(df_mismatch['active_time'].median())
        t_med = float(df_matched['active_time'].median())
        mismatch_summary = {
            "mismatch_n": int(len(df_mismatch)),
            "matched_n": int(len(df_matched)),
            "mismatch_active_time_med": m_med,
            "matched_active_time_med": t_med,
            "ratio_mismatch_over_matched": (m_med / t_med) if t_med > 0 else np.nan,
        }
        print(f"mismatch active_time median: {m_med:.1f}s")
        print(f"matched  active_time median: {t_med:.1f}s")

# 5) 结构化导出（便于论文引用/复现）
sensitivity_rows = [{
    "coverage_new": coverage_new,
    "coverage_old": coverage_old,
    "coverage_diff": coverage_new - coverage_old,
}]
if mismatch_summary:
    sensitivity_rows[0].update(mismatch_summary)
if rq2_rows:
    for r in rq2_rows:
        sensitivity_rows[0][f"IAA_med_{r['scope']}"] = r["IAA_med"]
if rq1_rows:
    for r in rq1_rows:
        sensitivity_rows[0][f"active_time_med_{r['scope']}_{r['condition']}"] = r["active_time_med"]
sensitivity_df = pd.DataFrame(sensitivity_rows)
display(sensitivity_df)
if "CONFIG" in globals():
    out_path = CONFIG["output_dir"] / "sensitivity_p0_1.csv"
    sensitivity_df.to_csv(out_path, index=False)
    print(f"[saved] {out_path}")
````

# Section 7: Stratification Tables (Table A/B)

- 按 |Δn_pairs| 分层
- 按 difficulty / model_issue 分层

## [表 A | 分层] 按 |Δn_pairs| 分层的覆盖率/效率/质量（Dula-Net风格）

**目的**：揭示结构差异子群的工作量与质量分布（mismatch 不是少数边角）。

**最少输出**：
- 分箱：$|\Delta n| = 0, 1, 2, >2$
- 每箱：样本数、coverage（layout_used）、active_time 中位数、boundary_rmse 中位数、IoU/IAA 中位数（可用时）

````python
# =============================================================================
# [表 A] 按 |Δn_pairs| 分层的覆盖率/效率/质量
# =============================================================================
# 数据集：df_inscope
# 目的：揭示结构差异子群的工作量与质量分布
# =============================================================================

def build_table_a(df: pd.DataFrame) -> pd.DataFrame:
    """按 |Δn_pairs| 分层统计。"""
    
    required = ["pred_n_pairs", "ann_n_pairs"]
    if not all(c in df.columns for c in required):
        print("⚠️  缺少 pred_n_pairs/ann_n_pairs，无法生成表A")
        return pd.DataFrame()
    
    df_work = df.copy()
    df_work["abs_delta_n_pairs"] = (
        to_numeric_safe(df_work["pred_n_pairs"]) - to_numeric_safe(df_work["ann_n_pairs"])
    ).abs()
    
    # 分箱：0, 1, 2, >2
    bins = [-0.5, 0.5, 1.5, 2.5, np.inf]
    labels = ["0", "1", "2", ">2"]
    df_work["delta_bin"] = pd.cut(
        df_work["abs_delta_n_pairs"], bins=bins, labels=labels, right=True, include_lowest=True
    )
    
    # 分组统计
    rows = []
    for bin_label in labels:
        g = df_work[df_work["delta_bin"] == bin_label]
        if len(g) == 0:
            continue
        
        row = {
            "|Δn_pairs|": bin_label,
            "n_rows": len(g),
            "n_tasks": g["task_id"].nunique() if "task_id" in g.columns else None,
            "coverage_layout": safe_mean(g.get("layout_used_clean", pd.Series(dtype=float)).map({True: 1, False: 0})),
            "active_time_med": safe_median(g.get("active_time", pd.Series(dtype=float))),
            "boundary_rmse_med": safe_median(g.get("boundary_rmse_px", pd.Series(dtype=float))),
            "iou_med": safe_median(g.get("iou", pd.Series(dtype=float))),
            "agree_med": safe_median(g.get("iou_to_others_median", pd.Series(dtype=float))),
        }
        rows.append(row)
    
    return pd.DataFrame(rows)

table_a = build_table_a(df_inscope)
if len(table_a) > 0:
    # 格式化显示
    table_a_display = table_a.copy()
    for col in ["coverage_layout"]:
        if col in table_a_display.columns:
            table_a_display[col] = table_a_display[col].apply(lambda x: f"{x:.1%}" if pd.notna(x) else "N/A")
    for col in ["active_time_med", "boundary_rmse_med"]:
        if col in table_a_display.columns:
            table_a_display[col] = table_a_display[col].apply(lambda x: f"{x:.1f}" if pd.notna(x) else "N/A")
    for col in ["iou_med", "agree_med"]:
        if col in table_a_display.columns:
            table_a_display[col] = table_a_display[col].apply(lambda x: f"{x:.3f}" if pd.notna(x) else "N/A")
    
    display(table_a_display)
    save_table(table_a, "Table_A_delta_n_pairs", category="tables", index=False)
````

## [表 B | 分层] 按 difficulty / model_issue 分层的效率与一致性

**目的**：展示主观难度/模型问题对效率与一致性的影响，避免只报整体均值。

**最少输出**：
- 分层：difficulty（如 low/med/high）与 model_issue（仅 semi 条件）
- 每层：样本数、active_time 中位数、IAA 中位数（可用时）

````python
# =============================================================================
# [表 B] 按 difficulty / model_issue 分层的效率与一致性
# =============================================================================
# 数据集：df_inscope
# 注意：difficulty/model_issue 是多选（用 ';' 拼接），需要 explode 处理
# =============================================================================

def explode_tags(df: pd.DataFrame, col: str, sep: str = ";") -> pd.DataFrame:
    """将多选字段展开为多行。"""
    if col not in df.columns:
        return df.assign(_tag=pd.Series(dtype=str))
    
    s = df[col].fillna("").astype(str)
    tags = s.str.split(sep)
    out = df.assign(_tag=tags).explode("_tag")
    out["_tag"] = out["_tag"].fillna("").astype(str).str.strip()
    out = out[out["_tag"] != ""]
    return out

def summarize_by_tag(df: pd.DataFrame, tag_col: str = "_tag") -> pd.DataFrame:
    """按 tag 统计。"""
    rows = []
    for tag in df[tag_col].unique():
        g = df[df[tag_col] == tag]
        rows.append({
            "tag": tag,
            "n_rows": len(g),
            "n_tasks": g["task_id"].nunique() if "task_id" in g.columns else None,
            "active_time_med": safe_median(g.get("active_time", pd.Series(dtype=float))),
            "agree_med": safe_median(g.get("iou_to_others_median", pd.Series(dtype=float))),
        })
    
    return pd.DataFrame(rows).sort_values("n_rows", ascending=False)

# B1: difficulty（全条件）
print("📊 Table B1: By Difficulty")
if "difficulty" in df_inscope.columns:
    diff_long = explode_tags(df_inscope, "difficulty")
    if len(diff_long) > 0:
        table_b_diff = summarize_by_tag(diff_long)
        display(table_b_diff)
        save_table(table_b_diff, "Table_B1_difficulty", category="tables", index=False)
    else:
        print("⚠️  difficulty 全为空")
else:
    print("⚠️  difficulty 字段不存在")

# B2: model_issue（仅 semi 条件）
print("\n📊 Table B2: By Model Issue (Semi-Auto only)")
if all(c in df_inscope.columns for c in ["model_issue", "condition_clean"]):
    semi_only = df_inscope[df_inscope["condition_clean"] == "semi"].copy()
    model_long = explode_tags(semi_only, "model_issue")
    if len(model_long) > 0:
        table_b_model = summarize_by_tag(model_long)
        display(table_b_model)
        save_table(table_b_model, "Table_B2_model_issue", category="tables", index=False)
    else:
        print("⚠️  model_issue 全为空")
else:
    print("⚠️  缺少 model_issue 或 condition_clean 字段")
````

## [内部图 Fig.D1 | 诊断] 按标注者耗时差异（log-y）

**用途**：效率异质性诊断与任务分配动机（内部分析图）。

**核心列**：`annotator_id`, `active_time`，（可选）`condition_clean`

**建议输出**：boxplot + stripplot（log-y），并给出每个标注者的 n（样本量）。

````python
# =============================================================================
# [内部图 Fig.D1] 按标注者耗时差异（log-y）
# =============================================================================

def plot_annotator_time_distribution(df: pd.DataFrame) -> plt.Figure:
    """绘制按标注者的耗时分布。"""
    
    if "active_time" not in df.columns or "annotator_id" not in df.columns:
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.text(0.5, 0.5, "缺少必要字段", ha="center", va="center")
        ax.axis("off")
        return fig
    
    df_plot = df.copy()
    df_plot["active_time"] = to_numeric_safe(df_plot["active_time"])
    df_plot = df_plot.dropna(subset=["active_time", "annotator_id"])
    
    # 按中位耗时排序
    order = df_plot.groupby("annotator_id")["active_time"].median().sort_values(ascending=False).index.tolist()
    order = order[:20]  # Top 20
    df_plot = df_plot[df_plot["annotator_id"].isin(order)]
    
    fig, ax = plt.subplots(figsize=(14, 6))
    
    # 是否按 condition 分色
    if "condition_clean" in df_plot.columns:
        df_plot["Condition"] = df_plot["condition_clean"].apply(condition_display)
        sns.boxplot(data=df_plot, x="annotator_id", y="active_time", hue="Condition",
                   order=order, ax=ax, palette={"Manual": "#4C72B0", "Semi-Auto": "#55A868"})
    else:
        sns.boxplot(data=df_plot, x="annotator_id", y="active_time", order=order, ax=ax)
    
    ax.set_yscale("log")
    ax.set_xlabel("Annotator ID")
    ax.set_ylabel("Active Time (seconds, log)")
    ax.set_title("Fig.D1: Active Time Distribution by Annotator")
    ax.tick_params(axis="x", rotation=45)
    
    plt.tight_layout()
    return fig

fig_d1 = plot_annotator_time_distribution(df_transparent)
save_figure(fig_d1, "Fig_D1_annotator_time", category="diagnostics")
plt.show()
````

## [内部图 Fig.D2 | 诊断] 任务×标注者热力图（分歧发现/定位）

**当前口径（2026-01-21）**：只在 notebook 中展示，不落盘保存图片文件。

**数据集**：建议用 `df_main`（主口径一致），也可对比 `df_transparent`
**核心列**：`task_id`, `annotator_id`, 以及一个主指标列（如 `iou` 或你论文指定的主指标）

**目的**：定位“哪个任务分歧最大 / 哪个标注者偏离更系统性”，用于标注指南迭代与审稿质疑回应。

**你要写的代码**：
1. pivot：index=task_id, columns=annotator_id, values=metric
2. 选择 top-N 最有信息的任务（如方差最大）
3. heatmap 绘制并标注缺失（NA）

> 如果未来确实需要导出图片，请单独写“导出版脚本/函数”，避免把 notebook 的诊断默认变成落盘流程。

````python
# =============================================================================
# [内部图 Fig.D2] 任务×标注者热力图（分歧发现）
# =============================================================================
# 当前口径：只在 notebook 中展示，不落盘保存图片文件
# =============================================================================

def plot_task_annotator_heatmap(
    df: pd.DataFrame,
    metric_col: str = "iou",
    top_n: int = 20,
    rank_by: str = "var",
) -> tuple:
    """绘制任务×标注者热力图。"""
    
    required = {"task_id", "annotator_id", metric_col}
    missing = required - set(df.columns)
    if missing:
        print(f"⚠️  缺少字段: {missing}")
        return None, None
    
    df_work = df[["task_id", "annotator_id", metric_col]].copy()
    df_work["task_id"] = df_work["task_id"].astype(str)
    df_work["annotator_id"] = df_work["annotator_id"].astype(str)
    df_work[metric_col] = to_numeric_safe(df_work[metric_col])
    
    # 构建 pivot 矩阵
    mat = df_work.pivot_table(
        index="task_id",
        columns="annotator_id",
        values=metric_col,
        aggfunc="mean"
    )
    
    # 筛选多标注任务
    ann_cnt = mat.notna().sum(axis=1)
    mat = mat.loc[ann_cnt >= 2]
    
    if mat.empty:
        print("⚠️  没有多标注任务")
        return None, None
    
    # 计算分歧分数并选择 top-N
    if rank_by == "range":
        score = mat.max(axis=1, skipna=True) - mat.min(axis=1, skipna=True)
    else:
        score = mat.var(axis=1, skipna=True)
    
    score = score.dropna()
    score = score[score > 0]
    
    if score.empty:
        print("⚠️  没有有效的分歧任务")
        return None, None
    
    top_tasks = score.sort_values(ascending=False).head(top_n).index
    mat_sel = mat.loc[top_tasks]
    
    # 按列均值排序
    col_order = mat_sel.mean(axis=0, skipna=True).sort_values(ascending=False).index
    mat_sel = mat_sel[col_order]
    
    # 绘图
    fig, ax = plt.subplots(figsize=(max(10, 0.8 * mat_sel.shape[1]), max(6, 0.5 * mat_sel.shape[0])))
    
    sns.heatmap(
        mat_sel,
        cmap="YlGnBu",
        annot=True,
        fmt=".2f",
        linewidths=0.3,
        linecolor="white",
        mask=mat_sel.isna(),
        vmin=0,
        vmax=1,
        cbar_kws={"label": metric_col},
        ax=ax,
    )
    ax.set_facecolor("#f0f0f0")
    ax.set_title(f"Fig.D2: Task x Annotator Heatmap (Top {len(mat_sel)} by {rank_by})")
    ax.set_xlabel("Annotator ID")
    ax.set_ylabel("Task ID")
    
    plt.tight_layout()
    
    # 返回分歧分数供进一步分析
    score_df = pd.DataFrame({
        "task_id": score.index,
        f"{rank_by}_score": score.values,
        "n_annotators": ann_cnt.loc[score.index].values,
    }).sort_values(f"{rank_by}_score", ascending=False)
    
    return fig, score_df

# 绘制热力图（仅展示，不保存）
iou_col = "iou" if "iou" in df_main.columns else "iou_corner"
fig_d2, divergence_df = plot_task_annotator_heatmap(df_main, metric_col=iou_col, top_n=15)

if fig_d2:
    plt.show()
    print("\n📋 Top 10 高分歧任务:")
    display(divergence_df.head(10))
    # 注意：不保存图片，仅 notebook 展示
````

## [附录/内部图 Fig.D3 | 审计] Mixed-scope 任务清单与投票分布（任务级）

**当前口径**：默认仅在 notebook 展示（不强制保存图片）。

**数据集**：`df_transparent`（需要按 `task_id` 聚合；CSV 已包含任务级 scope 投票字段可直接用）
**核心列**：`task_id`, `task_scope_n_total`, `task_scope_n_in_scope`, `task_scope_n_oos`, `task_scope_n_unknown`, `task_scope_is_mixed`

**目的**：找出“同一任务不同标注者对 scope 判定不一致”的样本，这是质量控制/标注指南迭代的关键证据。

**你要写的代码**：
1. 任务级筛选：`task_scope_is_mixed == True`
2. 输出 task 列表（附上每个 task 的 scope 投票分布）
3. （可选）画一个投票分布的条形图/堆叠条形图做直觉展示

````python
# =============================================================================
# [附录/内部图 Fig.D3] Mixed-scope 任务清单与投票分布
# =============================================================================
# 当前口径：默认仅在 notebook 展示
# =============================================================================

def analyze_mixed_scope_tasks(df: pd.DataFrame) -> tuple:
    """分析 mixed-scope 任务。"""
    
    if "task_scope_is_mixed" not in df.columns:
        print("⚠️  缺少 task_scope_is_mixed 字段")
        return None, None
    
    # 筛选 mixed-scope 任务
    mixed_tasks = df[df["task_scope_is_mixed"] == True].copy()
    
    if len(mixed_tasks) == 0:
        print("✅ 没有 mixed-scope 任务")
        return None, None
    
    # 按任务聚合
    scope_cols = ["task_scope_n_total", "task_scope_n_in_scope", "task_scope_n_oos", "task_scope_n_unknown"]
    available_cols = [c for c in scope_cols if c in mixed_tasks.columns]
    
    if "task_id" in mixed_tasks.columns:
        task_summary = mixed_tasks.groupby("task_id").first()[available_cols].reset_index()
    else:
        task_summary = mixed_tasks[available_cols].drop_duplicates()
    
    # 绘图
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Panel A: 投票分布堆叠条形图
    ax1 = axes[0]
    if all(c in task_summary.columns for c in ["task_scope_n_in_scope", "task_scope_n_oos"]):
        plot_data = task_summary.set_index("task_id")[["task_scope_n_in_scope", "task_scope_n_oos"]]
        if "task_scope_n_unknown" in task_summary.columns:
            plot_data["task_scope_n_unknown"] = task_summary.set_index("task_id")["task_scope_n_unknown"]
        
        plot_data.plot(kind="barh", stacked=True, ax=ax1,
                      color={"task_scope_n_in_scope": "#55A868", 
                             "task_scope_n_oos": "#C44E52",
                             "task_scope_n_unknown": "#999999"})
        ax1.set_xlabel("Vote Count")
        ax1.set_ylabel("Task ID")
        ax1.set_title(f"Panel A: Scope Votes (n={len(task_summary)} mixed tasks)")
        ax1.legend(["In-scope", "OOS", "Unknown"], title="Scope", loc="lower right")
    else:
        ax1.text(0.5, 0.5, "缺少投票字段", ha="center", va="center")
        ax1.axis("off")
    
    # Panel B: 统计摘要
    ax2 = axes[1]
    stats_text = f"📋 Mixed-Scope Tasks Summary\n{'=' * 35}\n\n"
    stats_text += f"总计: {len(task_summary)} 个任务\n\n"
    
    if "task_scope_n_total" in task_summary.columns:
        avg_annotators = task_summary["task_scope_n_total"].mean()
        stats_text += f"平均标注者数: {avg_annotators:.1f}\n"
    
    stats_text += "\n这些任务需要人工复核以确定正确的 scope 判定。"
    
    ax2.text(0.05, 0.95, stats_text, transform=ax2.transAxes, fontsize=11,
            verticalalignment="top", fontfamily="monospace",
            bbox=dict(boxstyle="round", facecolor="lightyellow", alpha=0.8))
    ax2.axis("off")
    ax2.set_title("Panel B: Summary")
    
    plt.suptitle("Fig.D3: Mixed-Scope Task Audit", fontsize=14, fontweight="bold", y=1.02)
    plt.tight_layout()
    
    return fig, task_summary

fig_d3, mixed_task_df = analyze_mixed_scope_tasks(df_transparent)
if fig_d3:
    plt.show()
    print("\n📋 Mixed-scope 任务清单:")
    display(mixed_task_df)
````

### [附录/内部表 Table.A1 | 审计] 低质量/异常样本清单（规则写死，可复现）

**目的**
- 给出“可复现的反例清单”：哪些样本在主分析口径下表现差、或存在明显异常（例如耗时极高、指标越界、scope/字段不一致等）。
- 该表用于审计/追踪与复现，不用于主文显著性结论。

**输入数据**
- 使用 `df_main`（统一主分析口径：in-scope 且（可选）layout_used=True）。
- 涉及某一指标（例如 IoU / boundary_rmse_px）时，必须额外加 `.notna()`：NA 表示“不可计算/缺失”，不是 0。

**建议列（每行一个 task×annotator）**
- `task_id, annotator_id, condition_clean`
- `scope_clean, scope_missing, is_oos_clean`
- `layout_used_clean, layout_gate_reason`
- `active_time`
- 指标：`iou`（或你论文定义的主 IoU 字段）、`boundary_rmse_px`、`pointwise_rmse`（若存在）
- 可靠性（可选）：`iou_to_consensus_loo, iou_to_others_median`
- 异常标记：`flag_*` 与 `main_exclude_reason`

**排序建议（只在该指标可计算时排序）**
- “最差 IoU” → `iou` 升序（先 `.notna()`）
- “最差边界” → `boundary_rmse_px` 降序（先 `.notna()`）
- “高耗时低质量” → `active_time` 降序，且要求 `active_time` 与 `iou` 均可计算

```python
# TODO: 从 df_main 中筛选/排序，生成一个审计用 DataFrame（不做可视化）
```


````python
# =============================================================================
# [附录/内部表 Table.A1] 低质量/异常样本清单（审计/复现）
# =============================================================================
# 数据集：df_main
# 目的：可复现的反例清单
# =============================================================================

def build_anomaly_audit_table(df: pd.DataFrame, top_k: int = 20) -> pd.DataFrame:
    """构建异常样本审计表。"""
    
    # 定义输出列
    id_cols = ["task_id", "annotator_id", "condition_clean"]
    scope_cols = ["scope_clean", "is_oos_clean"]
    metric_cols = ["iou", "iou_corner", "boundary_rmse_px", "active_time"]
    reliability_cols = ["iou_to_consensus_loo", "iou_to_others_median"]
    gate_cols = ["layout_used_clean", "layout_gate_reason", "main_exclude_reason"]
    
    all_cols = id_cols + scope_cols + metric_cols + reliability_cols + gate_cols
    available_cols = [c for c in all_cols if c in df.columns]
    
    df_audit = df[available_cols].copy()
    
    # 转换数值列
    for col in metric_cols + reliability_cols:
        if col in df_audit.columns:
            df_audit[col] = to_numeric_safe(df_audit[col])
    
    anomalies = []
    
    # 类别1: 最低 IoU
    iou_col = "iou" if "iou" in df_audit.columns else "iou_corner"
    if iou_col in df_audit.columns:
        low_iou = df_audit.dropna(subset=[iou_col]).nsmallest(top_k, iou_col)
        low_iou["anomaly_type"] = f"Low {iou_col}"
        anomalies.append(low_iou)
    
    # 类别2: 最高 boundary_rmse
    if "boundary_rmse_px" in df_audit.columns:
        high_rmse = df_audit.dropna(subset=["boundary_rmse_px"]).nlargest(top_k, "boundary_rmse_px")
        high_rmse["anomaly_type"] = "High Boundary RMSE"
        anomalies.append(high_rmse)
    
    # 类别3: 最高 active_time
    if "active_time" in df_audit.columns:
        high_time = df_audit.dropna(subset=["active_time"]).nlargest(top_k, "active_time")
        high_time["anomaly_type"] = "High Active Time"
        anomalies.append(high_time)
    
    # 类别4: 最低一致性
    if "iou_to_others_median" in df_audit.columns:
        low_agree = df_audit.dropna(subset=["iou_to_others_median"]).nsmallest(top_k, "iou_to_others_median")
        low_agree["anomaly_type"] = "Low Agreement"
        anomalies.append(low_agree)
    
    if not anomalies:
        return pd.DataFrame()
    
    # 合并并去重
    result = pd.concat(anomalies, ignore_index=True)
    
    # 按 task_id + annotator_id 去重，保留第一个 anomaly_type
    if "task_id" in result.columns and "annotator_id" in result.columns:
        result = result.drop_duplicates(subset=["task_id", "annotator_id"], keep="first")
    
    return result.sort_values("anomaly_type")

# 生成审计表
table_a1 = build_anomaly_audit_table(df_main, top_k=10)
if len(table_a1) > 0:
    print(f"📋 Table.A1: 异常样本审计表 (n={len(table_a1)})")
    display(table_a1)
    save_table(table_a1, "Table_A1_anomaly_audit", category="tables", index=False)
else:
    print("⚠️  无法生成异常样本表")
````

### I. 关于 IoU=NA 的语义（论文/审稿口径建议）
- NA 表示"不可计算/缺失"，不是 0 分：
  - 例如缺少预测、多边形退化、被 gate 掉的计算路径。
- 所有涉及 IoU 的排序/统计/绘图必须显式 `.notna()`；缺失率可作为透明性指标单独报告。

### J. 2026-01-19~01-22 重要变更记录

| 变更项 | 原逻辑 | 新逻辑 | 影响 |
